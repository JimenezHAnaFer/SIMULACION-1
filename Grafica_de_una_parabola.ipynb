```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(-10, 10, 100)
y = x**2

# Crear la gráfica
plt.plot(x, y, linestyle='-', color='b', label='Parábola: y = x^2')

# Etiquetas y título
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfica de una Parábola')
plt.legend()

# Mostrar la gráfica
plt.show()
```
