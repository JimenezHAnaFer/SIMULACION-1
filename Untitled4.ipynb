```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de theta de 0 a 2pi
theta = np.linspace(0, 2 * np.pi, 100)

# Calcular las coordenadas del círculo unitario
x = np.cos(theta)
y = np.sin(theta)

# Graficar el círculo
plt.plot(x, y)
plt.axis('equal')  # Asegura que los ejes tengan la misma escala
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Generar puntos (dardos)
lista = []

for i in range(10):
    x = random.uniform(0, 1)
    y = random.uniform(0, 1)
    lista.append((x, y))
    plt.scatter(x, y, s=5)  # Graficar los puntos (dardos)

# Dibujar un cuarto del círculo (cuadrante 1)
theta = np.linspace(0, np.pi/2, 100)
x = np.cos(theta)
y = np.sin(theta)
plt.plot(x, y, color='red')  # Graficar el cuarto de círculo

# Ajustar los ejes para que todo se vea bien
plt.axis('equal')  # Asegura que los ejes tengan la misma escala
plt.xlim(0, 1)  # Establece el rango en el eje x
plt.ylim(0, 1)  # Establece el rango en el eje y
plt.show()

```
