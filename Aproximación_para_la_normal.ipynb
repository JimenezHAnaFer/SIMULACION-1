```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd

#generamos la funcion para crear x

lista=[]
def aproximacion (n):
  for _ in range (n):
    R=rd.random()
    x=(R**0.135-(1-R)**0.135)/0.1975
    lista.append(x)
  return lista

experimentos=aproximacion(1000000)


print('la media es', np.mean(experimentos))
print('la varianza es', np.var(experimentos))

plt.hist(experimentos, bins=100,  color='teal')
plt.grid(True, linestyle='--', alpha=0.7)
plt.title('Histograma de la distribuci√≥n')
plt.show()
```
