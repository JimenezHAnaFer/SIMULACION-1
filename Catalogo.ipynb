## Uniforme



**Nomenclatura**



*   **Notación:** \( X \sim U(a, b) \) o \( X \sim Unif(a, b) \)
    *   "La variable aleatoria X sigue una distribución uniforme continua con parámetros \( a \) y \( b \)."

*   **Parámetros:**
    *   \( a \): Límite inferior del intervalo. \( a < b \).
    *   \( b \): Límite superior del intervalo.

*   **Soporte:** \( [a, b] \)
    *   El conjunto de valores que la variable aleatoria puede tomar. Es un intervalo continuo.

*   **Función de Densidad de Probabilidad (FDP) - *Probability Density Function (PDF)*:**
    *   \( f(x) = \begin{cases} \frac{1}{b - a} & \text{para } a \le x \le b \\ 0 & \text{para otro caso} \end{cases} \)

*   **Función de Distribución Acumulada (FDA) - *Cumulative Distribution Function (CDF)*:**
    *   \( F(x) = \begin{cases} 0 & \text{para } x < a \\ \frac{x - a}{b - a} & \text{para } a \le x \le b \\ 1 & \text{para } x > b \end{cases} \)

*   **Esperanza (Media):**
    $$ E(X) = \frac{a + b}{2} $$

*   **Varianza :**
    $$ Var(X) = \frac{(b - a)^2}{12} $$

*   **Desviación Estándar :**
    $$ \sigma = \sqrt{Var(X)} = \frac{b - a}{\sqrt{12}} $$

*   **Moda:**
    *   Cualquier valor en el intervalo \( [a, b] \).

*   **Mediana :**
    $$ \text{Mediana} = \frac{a + b}{2} $$

##Triangular


**Nomenclatura**



*   **Notación:** \( X \sim \text{Triangular}(a, c, b) \)
    *   "La variable aleatoria X sigue una distribución triangular continua con parámetros \( a \), \( c \) y \( b \)."

*   **Parámetros:**
    *   \( a \): Límite inferior o valor mínimo.
    *   \( c \): Valor modal o moda (la cumbre del triángulo). \( a \le c \le b \).
    *   \( b \): Límite superior o valor máximo. \( a < b \).

*   **Soporte:** \( [a, b] \)
    *   El conjunto de valores que la variable aleatoria puede tomar.

*   **Función de Densidad de Probabilidad (FDP) - *Probability Density Function (PDF)*:**
    *   \( f(x) = \begin{cases} 0 & \text{para } x < a \\ \frac{2(x-a)}{(b-a)(c-a)} & \text{para } a \le x < c \\ \frac{2}{b-a} & \text{para } x = c \\ \frac{2(b-x)}{(b-a)(b-c)} & \text{para } c < x \le b \\ 0 & \text{para } x > b \end{cases} \)
    *   Representa la altura del triángulo en el punto \( x \).

*   **Función de Distribución Acumulada (FDA) - *Cumulative Distribution Function (CDF)*:**
    *   \( F(x) = \begin{cases} 0 & \text{para } x < a \\ \frac{(x-a)^2}{(b-a)(c-a)} & \text{para } a \le x < c \\ 1 - \frac{(b-x)^2}{(b-a)(b-c)} & \text{para } c \le x \le b \\ 1 & \text{para } x > b \end{cases} \)
    *   Representa la probabilidad de que la variable aleatoria sea menor o igual a un valor dado \( x \).

*   **Esperanza (Media) :**
    $$ E(X) = \frac{a + c + b}{3} $$

*   **Varianza:**
    $$ Var(X) = \frac{a^2 + b^2 + c^2 - ab - ac - bc}{18} $$

*   **Moda :**
    $$ \text{Moda} = c $$
    *   El valor con la mayor densidad de probabilidad.

*   **Mediana :**
    *   Si \( c \ge \frac{a+b}{2} \), $$ \text{Mediana} = a + \sqrt{(b-a)(c-a)/2} $$
    *   Si \( c \le \frac{a+b}{2} \), $$ \text{Mediana} = b - \sqrt{(b-a)(b-c)/2} $$
    *   El valor que divide el área del triángulo en dos partes iguales.

##m-Erlang

**Nomenclatura**

*   **Notación:** \( X \sim \text{Erlang}(m, \lambda) \) o \( X \sim \text{Gamma}(m, \lambda) \) con \( m \) entero
    *   "La variable aleatoria X sigue una distribución Erlang con parámetros \( m \) y \( \lambda \)."

*   **Parámetros:**
    *   \( m \): Parámetro de forma (entero positivo). Número de etapas exponenciales independientes.
    *   \( \lambda \): Parámetro de tasa (positivo). Tasa de ocurrencia en cada etapa. \( \lambda > 0 \).

*   **Soporte:** \( [0, \infty) \)
    *   El conjunto de valores que la variable aleatoria puede tomar (valores no negativos).

*   **Función de Densidad de Probabilidad (FDP) - *Probability Density Function (PDF)*:**
    *   \( f(x) = \begin{cases}
    \frac{\lambda^m x^{m-1} e^{-\lambda x}}{(m-1)!} & \text{para } x \ge 0 \\
    0 & \text{para } x < 0
    \end{cases} \)
    *   Donde \( (m-1)! \) es el factorial de \( m-1 \).

*   **Función de Distribución Acumulada (FDA) - *Cumulative Distribution Function (CDF)*:**
    *   \( F(x) = \begin{cases}
    1 - e^{-\lambda x} \sum_{k=0}^{m-1} \frac{(\lambda x)^k}{k!} & \text{para } x \ge 0 \\
    0 & \text{para } x < 0
    \end{cases} \)
    *   Representa la probabilidad acumulada hasta el valor \( x \).

*   **Esperanza (Media):**
    $$ E(X) = \frac{m}{\lambda} $$

*   **Varianza:**
    $$ Var(X) = \frac{m}{\lambda^2} $$

*   **Moda:**
    $$ \text{Moda} = \begin{cases}
    \frac{m-1}{\lambda} & \text{para } m \ge 1 \\
    0 & \text{para } m = 1
    \end{cases} $$
    *   El valor con mayor densidad de probabilidad.

*   **Mediana:**
    *   No tiene fórmula cerrada simple. Se obtiene resolviendo:
    $$ 1 - e^{-\lambda x} \sum_{k=0}^{m-1} \frac{(\lambda x)^k}{k!} = 0.5 $$
    *   El valor que divide la distribución en dos partes de igual probabilidad.

*   **Propiedad Aditiva:**
    *   Si \( X_i \sim \text{Erlang}(m_i, \lambda) \) son independientes, entonces:
    $$ \sum_{i=1}^n X_i \sim \text{Erlang}\left(\sum_{i=1}^n m_i, \lambda\right) $$

*   **Relación con la Exponencial:**
    *   Cuando \( m = 1 \), se reduce a una distribución exponencial con tasa \( \lambda \).

##Gamma

**Nomenclatura**

*   **Notación:** \( X \sim \text{Gamma}(\alpha, \beta) \) o \( X \sim \Gamma(\alpha, \beta) \)
    *   "La variable aleatoria X sigue una distribución Gamma con parámetros \(\alpha\) y \(\beta\)."

*   **Parámetros:**
    *   \(\alpha\): Parámetro de forma (positivo). Controla la forma de la distribución. \(\alpha > 0\)
    *   \(\beta\): Parámetro de escala (positivo). Controla la dispersión. \(\beta > 0\)
    *   (Nota alternativa: Algunos textos usan \(\lambda = 1/\beta\) como parámetro de tasa)

*   **Soporte:** \( [0, \infty) \)
    *   El conjunto de valores que la variable aleatoria puede tomar (valores no negativos).

*   **Función de Densidad de Probabilidad (FDP) - *Probability Density Function (PDF)*:**
    *   \( f(x) = \begin{cases}
    \frac{x^{\alpha-1} e^{-x/\beta}}{\beta^\alpha \Gamma(\alpha)} & \text{para } x \ge 0 \\
    0 & \text{para } x < 0
    \end{cases} \)
    *   Donde \(\Gamma(\alpha)\) es la función Gamma.

*   **Función de Distribución Acumulada (FDA) - *Cumulative Distribution Function (CDF)*:**
    *   \( F(x) = \begin{cases}
    \frac{\gamma(\alpha, x/\beta)}{\Gamma(\alpha)} & \text{para } x \ge 0 \\
    0 & \text{para } x < 0
    \end{cases} \)
    *   Donde \(\gamma(\alpha, z)\) es la función Gamma incompleta inferior.

*   **Esperanza (Media):**
    $$ E(X) = \alpha \beta $$

*   **Varianza:**
    $$ Var(X) = \alpha \beta^2 $$

*   **Moda:**
    $$ \text{Moda} = \begin{cases}
    (\alpha-1)\beta & \text{para } \alpha \ge 1 \\
    0 & \text{para } \alpha < 1
    \end{cases} $$
    *   El valor con mayor densidad de probabilidad.

*   **Mediana:**
    *   No tiene fórmula cerrada simple. Se obtiene resolviendo:
    $$ F(x) = 0.5 $$

*   **Función Generadora de Momentos:**
    $$ M_X(t) = (1 - \beta t)^{-\alpha} \quad \text{para } t < 1/\beta $$

*   **Propiedades Especiales:**
    1. **Caso Exponencial:** Cuando \(\alpha = 1\), se reduce a una distribución exponencial.
    2. **Caso Erlang:** Cuando \(\alpha\) es entero, se convierte en distribución Erlang.
    3. **Caso Chi-cuadrado:** Cuando \(\alpha = \nu/2\) y \(\beta = 2\), se convierte en \(\chi^2(\nu)\).
    4. **Propiedad Aditiva:** La suma de Gammas independientes con mismo \(\beta\) es Gamma.

*   **Relación con Otras Distribuciones:**
    *   Es la distribución conjugada prior para la precisión (inverso de la varianza) en distribuciones normales.
    *   Modela tiempos de espera para múltiples eventos en procesos de Poisson.

##Weibull

**Nomenclatura**

*   **Notación:** \( X \sim \text{Weibull}(k, \lambda) \)
    *   "La variable aleatoria X sigue una distribución Weibull con parámetros de forma \( k \) y escala \( \lambda \)."

*   **Parámetros:**
    *   \( k \): Parámetro de forma (positivo). Controla el comportamiento de la tasa de fallo. \( k > 0 \)
    *   \( \lambda \): Parámetro de escala (positivo). \( \lambda > 0 \)
    *   (Alternativamente: Algunos textos usan \( \alpha = \lambda^k \))

*   **Soporte:** \( [0, \infty) \)
    *   Valores no negativos (modela tiempos de vida/duración).

*   **Función de Densidad de Probabilidad (FDP):**
    *   \( f(x) = \begin{cases}
    \frac{k}{\lambda} \left( \frac{x}{\lambda} \right)^{k-1} e^{-(x/\lambda)^k} & \text{para } x \geq 0 \\
    0 & \text{para } x < 0
    \end{cases} \)

*   **Función de Distribución Acumulada (FDA):**
    *   \( F(x) = \begin{cases}
    1 - e^{-(x/\lambda)^k} & \text{para } x \geq 0 \\
    0 & \text{para } x < 0
    \end{cases} \)

*   **Función de Supervivencia:**
    $$ S(x) = e^{-(x/\lambda)^k} \quad \text{para } x \geq 0 $$

*   **Función de Riesgo (Tasa de Fallo):**
    $$ h(x) = \frac{k}{\lambda} \left( \frac{x}{\lambda} \right)^{k-1} $$

*   **Esperanza (Media):**
    $$ E(X) = \lambda \, \Gamma\left(1 + \frac{1}{k}\right) $$

*   **Varianza:**
    $$ Var(X) = \lambda^2 \left[ \Gamma\left(1 + \frac{2}{k}\right) - \left( \Gamma\left(1 + \frac{1}{k}\right) \right)^2 \right] $$

*   **Moda:**
    $$ \text{Moda} = \begin{cases}
    \lambda \left( \frac{k-1}{k} \right)^{1/k} & \text{para } k \geq 1 \\
    0 & \text{para } 0 < k < 1
    \end{cases} $$

*   **Mediana:**
    $$ \text{Mediana} = \lambda (\ln 2)^{1/k} $$

*   **Casos Especiales:**
    1. **Exponencial:** Cuando \( k = 1 \), se reduce a distribución exponencial.
    2. **Rayleigh:** Cuando \( k = 2 \), se convierte en distribución Rayleigh.
    3. **Determinística:** Cuando \( k \to \infty \), converge a \( \lambda \).

*   **Aplicaciones Típicas:**
    *   Análisis de supervivencia y fiabilidad
    *   Modelado de tiempos de fallo en ingeniería
    *   Datos meteorológicos (distribución de vientos)

##Normal



**Nomenclatura**

*   **Notación:** \( X \sim N(\mu, \sigma^2) \)
    *   "La variable aleatoria X sigue una distribución normal con media \(\mu\) y varianza \(\sigma^2\)."

*   **Parámetros:**
    *   \(\mu\): Parámetro de ubicación (media). Puede ser cualquier valor real.
    *   \(\sigma\): Parámetro de escala (desviación estándar). \(\sigma > 0\).

*   **Soporte:** \( (-\infty, \infty) \)
    *   Todos los números reales.

*   **Función de Densidad de Probabilidad (FDP):**
    $$ f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2} $$

*   **Función de Distribución Acumulada (FDA):**
    *   No tiene forma cerrada. Se expresa mediante:
    $$ F(x) = \Phi\left(\frac{x-\mu}{\sigma}\right) $$
    *   Donde \(\Phi\) es la FDA de la normal estándar \(N(0,1)\).

*   **Esperanza (Media):**
    $$ E(X) = \mu $$

*   **Varianza:**
    $$ Var(X) = \sigma^2 $$

*   **Desviación Estándar:**
    $$ \sigma = \sqrt{Var(X)} $$

*   **Moda:**
    $$ \text{Moda} = \mu $$
    *   Coincide con la media y mediana.

*   **Mediana:**
    $$ \text{Mediana} = \mu $$

*   **Simetría:**
    $$ \text{Sesgo} = 0 $$
    *   La distribución es perfectamente simétrica.

*   **Curtosis:**
    $$ \text{Curtosis} = 3 $$
    *   La curtosis excesiva es 0 (comparada con la normal).

*   **Estandarización:**
    $$ Z = \frac{X - \mu}{\sigma} \sim N(0, 1) $$

*   **Propiedad Aditiva:**
    *   Si \(X_i \sim N(\mu_i, \sigma_i^2)\) son independientes:
    $$ \sum_{i=1}^n X_i \sim N\left(\sum_{i=1}^n \mu_i, \sum_{i=1}^n \sigma_i^2\right) $$

*   **Teorema del Límite Central:**
    *   La suma de \(n\) variables aleatorias i.i.d. converge a \(N(n\mu, n\sigma^2)\) cuando \(n \to \infty\).

*   **Aplicaciones:**
    1. Modelado de errores de medición
    2. Fenómenos naturales (alturas, pesos)
    3. Inferencia estadística
    4. Finanzas (movimientos de precios)

#Codigo de grafica

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson # Importamos la distribución de Poisson

# ── Parámetros de la distribución de Poisson Discreta ───────────────────
lambda_param = 3.5  # Tasa promedio de eventos por intervalo (lambda) (lambda > 0)

# ── Datos para la gráfica ───────────────────────────────────────────────
# Los posibles resultados son el número de eventos (0, 1, 2, ...).
# Como no podemos graficar hasta el infinito, elegimos un número razonable de valores.
# Un buen rango para mostrar suele ser alrededor de lambda +/- 4*sqrt(lambda).
max_k = int(lambda_param + 4 * np.sqrt(lambda_param)) + 2
if max_k < 10: # Asegurar que al menos se muestren algunos valores
    max_k = 10

x = np.arange(0, max_k + 1) # Valores posibles para el número de eventos: 0, 1, 2, ..., max_k

# Calculamos la probabilidad para cada valor k (número de eventos) usando la PMF
# P(X=k | lambda)
pmf = poisson.pmf(x, lambda_param)

# La suma total de probabilidades hasta infinito es 1.
# Aquí calculamos la suma de las probabilidades graficadas hasta max_k.
total_prob_displayed = pmf.sum()

# ── Gráfica ─────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(10, 6))

# Usamos 'stem' para graficar la PMF de una distribución discreta.
markerline, stemlines, baseline = ax.stem(x, pmf, basefmt=' ')
plt.setp(stemlines, linewidth=2.5, color='tab:orange') # Puedes cambiar el color
plt.setp(markerline, markersize=8, marker='o', color='tab:orange')
baseline.set_visible(False)

# Etiquetas y título
ax.set_title(f'Distribución de Poisson (λ={lambda_param:.2f})', fontsize=14, weight='bold')
ax.set_xlabel('Número de Eventos (k)', fontsize=12)
ax.set_ylabel('P(X=k)', fontsize=12)

# Anotación de la suma de probabilidades (para los valores mostrados)
ax.text(max_k * 0.6, max(pmf)*0.9, # Ajusta la posición del texto
        f'Suma de prob. (hasta k={max_k}) = {total_prob_displayed:.4f}',
        ha='left', fontsize=10, style='italic')

# Límites y rejilla
ax.set_xlim(-0.5, max_k + 0.5) # Ajusta los límites del eje x
ax.set_ylim(0, max(pmf)*1.1)
ax.grid(axis='y', alpha=0.3)

# Asegura que los ticks del eje x sean enteros.
ax.set_xticks(x)

plt.tight_layout()
plt.show()
```
