##m-Erlang

**Nomenclatura**

**Gr√°fica**

**Pseudocodigo**

**C√≥digo**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define los par√°metros de la distribuci√≥n Erlang
k = 2  # N√∫mero de eventos (m)
lambda_param = 1  # Tasa de ocurrencia de eventos (lambda)

# Genera valores de x para la gr√°fica
x = np.linspace(0, 10, 100)

# Calcula la funci√≥n de densidad de probabilidad (PDF)
pdf = stats.erlang.pdf(x, k, scale=1/lambda_param)

# Crea la gr√°fica
plt.plot(x, pdf)
plt.title('Distribuci√≥n m-Erlang')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

##Gamma

**Nomenclatura**

**Gr√°fica**

**Pseudocodigo**

**C√≥digo**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define los par√°metros de la distribuci√≥n Gamma
alpha = 2  # Par√°metro de forma
beta = 1  # Par√°metro de tasa (inversa de la escala)

# Genera valores de x para la gr√°fica
x = np.linspace(0, 10, 100)

# Calcula la funci√≥n de densidad de probabilidad (PDF)
pdf = stats.gamma.pdf(x, a=alpha, scale=1/beta)

# Crea la gr√°fica
plt.plot(x, pdf)
plt.title('Distribuci√≥n Gamma')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

##Weibull

**Nomenclatura**

**Gr√°fica**

**Pseudocodigo**

**C√≥digo**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define los par√°metros de la distribuci√≥n Weibull
c = 2  # Par√°metro de forma
loc = 0  # Par√°metro de localizaci√≥n (desplazamiento)
scale = 1  # Par√°metro de escala

# Genera valores de x para la gr√°fica
x = np.linspace(0, 5, 100)

# Calcula la funci√≥n de densidad de probabilidad (PDF)
pdf = stats.weibull_min.pdf(x, c, loc=loc, scale=scale)

# Crea la gr√°fica
plt.plot(x, pdf)
plt.title('Distribuci√≥n Weibull')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

##Normal



**Nomenclatura**

**Gr√°fica**

**Pseudocodigo**

**C√≥digo**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define los par√°metros de la distribuci√≥n normal
mu = 0  # Media
sigma = 1  # Desviaci√≥n est√°ndar

# Genera valores de x para la gr√°fica
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

# Calcula la funci√≥n de densidad de probabilidad (PDF)
pdf = stats.norm.pdf(x, mu, sigma)

# Crea la gr√°fica
plt.plot(x, pdf)
plt.title('Distribuci√≥n Normal')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

##Log-normal

**Nomenclatura**

**Gr√°fica**

**Pseudocodigo**

**C√≥digo**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define los par√°metros de la distribuci√≥n log-normal
mu = 0  # Media del logaritmo natural de la variable
sigma = 1  # Desviaci√≥n est√°ndar del logaritmo natural de la variable

# Genera valores de x para la gr√°fica
x = np.linspace(0, 10, 100)  # Rango de valores para la variable original

# Calcula la funci√≥n de densidad de probabilidad (PDF)
pdf = stats.lognorm.pdf(x, s=sigma, scale=np.exp(mu))

# Crea la gr√°fica
plt.plot(x, pdf)
plt.title('Distribuci√≥n Log-normal')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

##Beta

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define los par√°metros de la distribuci√≥n Beta
a = 2  # Par√°metro de forma 1
b = 5  # Par√°metro de forma 2

# Genera valores de x para la gr√°fica
x = np.linspace(0, 1, 100)  # La distribuci√≥n Beta est√° definida en el intervalo [0, 1]

# Calcula la funci√≥n de densidad de probabilidad (PDF)
pdf = stats.beta.pdf(x, a, b)

# Crea la gr√°fica
plt.plot(x, pdf)
plt.title('Distribuci√≥n Beta')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

##Log-log√≠stica

**Nomenclatura**

**Gr√°fica**

**Pseudocodigo**

**C√≥digo**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define los par√°metros de la distribuci√≥n log-log√≠stica
alpha = 1  # Par√°metro de escala
beta = 1  # Par√°metro de forma

# Genera valores de x para la gr√°fica
x = np.linspace(0, 10, 100)

# Calcula la funci√≥n de densidad de probabilidad (PDF)
pdf = stats.fisk.pdf(x, c=beta, scale=alpha)  # La distribuci√≥n fisk es equivalente a la log-log√≠stica

# Crea la gr√°fica
plt.plot(x, pdf)
plt.title('Distribuci√≥n Log-log√≠stica')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

##Johnson


**Nomenclatura**

**Gr√°fica**

**Pseudocodigo**

**C√≥digo**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define los par√°metros de la distribuci√≥n Johnson SU
a = 2  # Par√°metro de forma 1
b = 3  # Par√°metro de forma 2
loc = 0  # Par√°metro de localizaci√≥n
scale = 1  # Par√°metro de escala

# Genera valores de x para la gr√°fica
x = np.linspace(-5, 5, 100)

# Calcula la funci√≥n de densidad de probabilidad (PDF)
pdf = stats.johnsonsu.pdf(x, a, b, loc=loc, scale=scale)

# Crea la gr√°fica
plt.plot(x, pdf)
plt.title('Distribuci√≥n Johnson SU')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

#Discretas

**Nomenclatura**

**Gr√°fica**

**Pseudocodigo**

**C√≥digo**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define los par√°metros de la distribuci√≥n uniforme discreta
a = 1  # L√≠mite inferior
b = 6  # L√≠mite superior

# Genera valores de x para la gr√°fica
x = np.arange(a, b + 1)

# Calcula la funci√≥n de masa de probabilidad (PMF)
pmf = np.full_like(x, 1 / (b - a + 1), dtype=float)

# Crea la gr√°fica
plt.stem(x, pmf, basefmt=' ')  # Usa stem para gr√°ficos discretos
plt.title('Distribuci√≥n Uniforme Discreta')
plt.xlabel('x')
plt.ylabel('Probabilidad')
plt.show()
```
