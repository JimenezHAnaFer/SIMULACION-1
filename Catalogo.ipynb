##m-Erlang

**Nomenclatura**

**Gráfica**

**Pseudocodigo**

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define los parámetros de la distribución Erlang
k = 2  # Número de eventos (m)
lambda_param = 1  # Tasa de ocurrencia de eventos (lambda)

# Genera valores de x para la gráfica
x = np.linspace(0, 10, 100)

# Calcula la función de densidad de probabilidad (PDF)
pdf = stats.erlang.pdf(x, k, scale=1/lambda_param)

# Crea la gráfica
plt.plot(x, pdf)
plt.title('Distribución m-Erlang')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

##Gamma

**Nomenclatura**

**Gráfica**

**Pseudocodigo**

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define los parámetros de la distribución Gamma
alpha = 2  # Parámetro de forma
beta = 1  # Parámetro de tasa (inversa de la escala)

# Genera valores de x para la gráfica
x = np.linspace(0, 10, 100)

# Calcula la función de densidad de probabilidad (PDF)
pdf = stats.gamma.pdf(x, a=alpha, scale=1/beta)

# Crea la gráfica
plt.plot(x, pdf)
plt.title('Distribución Gamma')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

##Weibull

**Nomenclatura**

**Gráfica**

**Pseudocodigo**

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define los parámetros de la distribución Weibull
c = 2  # Parámetro de forma
loc = 0  # Parámetro de localización (desplazamiento)
scale = 1  # Parámetro de escala

# Genera valores de x para la gráfica
x = np.linspace(0, 5, 100)

# Calcula la función de densidad de probabilidad (PDF)
pdf = stats.weibull_min.pdf(x, c, loc=loc, scale=scale)

# Crea la gráfica
plt.plot(x, pdf)
plt.title('Distribución Weibull')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

##Normal



**Nomenclatura**

**Gráfica**

**Pseudocodigo**

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define los parámetros de la distribución normal
mu = 0  # Media
sigma = 1  # Desviación estándar

# Genera valores de x para la gráfica
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

# Calcula la función de densidad de probabilidad (PDF)
pdf = stats.norm.pdf(x, mu, sigma)

# Crea la gráfica
plt.plot(x, pdf)
plt.title('Distribución Normal')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

##Log-normal

**Nomenclatura**

**Gráfica**

**Pseudocodigo**

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define los parámetros de la distribución log-normal
mu = 0  # Media del logaritmo natural de la variable
sigma = 1  # Desviación estándar del logaritmo natural de la variable

# Genera valores de x para la gráfica
x = np.linspace(0, 10, 100)  # Rango de valores para la variable original

# Calcula la función de densidad de probabilidad (PDF)
pdf = stats.lognorm.pdf(x, s=sigma, scale=np.exp(mu))

# Crea la gráfica
plt.plot(x, pdf)
plt.title('Distribución Log-normal')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define los parámetros de la distribución Beta
a = 2  # Parámetro de forma 1
b = 5  # Parámetro de forma 2

# Genera valores de x para la gráfica
x = np.linspace(0, 1, 100)  # La distribución Beta está definida en el intervalo [0, 1]

# Calcula la función de densidad de probabilidad (PDF)
pdf = stats.beta.pdf(x, a, b)

# Crea la gráfica
plt.plot(x, pdf)
plt.title('Distribución Beta')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

##Log-logística

**Nomenclatura**

**Gráfica**

**Pseudocodigo**

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define los parámetros de la distribución log-logística
alpha = 1  # Parámetro de escala
beta = 1  # Parámetro de forma

# Genera valores de x para la gráfica
x = np.linspace(0, 10, 100)

# Calcula la función de densidad de probabilidad (PDF)
pdf = stats.fisk.pdf(x, c=beta, scale=alpha)  # La distribución fisk es equivalente a la log-logística

# Crea la gráfica
plt.plot(x, pdf)
plt.title('Distribución Log-logística')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

##Johnson

**Pseudocodigo**

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define los parámetros de la distribución Johnson SU
a = 2  # Parámetro de forma 1
b = 3  # Parámetro de forma 2
loc = 0  # Parámetro de localización
scale = 1  # Parámetro de escala

# Genera valores de x para la gráfica
x = np.linspace(-5, 5, 100)

# Calcula la función de densidad de probabilidad (PDF)
pdf = stats.johnsonsu.pdf(x, a, b, loc=loc, scale=scale)

# Crea la gráfica
plt.plot(x, pdf)
plt.title('Distribución Johnson SU')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

#Discretas

**Nomenclatura**

**Gráfica**

**Pseudocodigo**

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define los parámetros de la distribución uniforme discreta
a = 1  # Límite inferior
b = 6  # Límite superior

# Genera valores de x para la gráfica
x = np.arange(a, b + 1)

# Calcula la función de masa de probabilidad (PMF)
pmf = np.full_like(x, 1 / (b - a + 1), dtype=float)

# Crea la gráfica
plt.stem(x, pmf, basefmt=' ')  # Usa stem para gráficos discretos
plt.title('Distribución Uniforme Discreta')
plt.xlabel('x')
plt.ylabel('Probabilidad')
plt.show()
```
