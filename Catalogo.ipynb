## Uniforme



**Nomenclatura**



*   **Notación:** $ X \sim U(a, b) $ o $ X \sim Unif(a, b) $
    *   "La variable aleatoria X sigue una distribución uniforme continua con parámetros \( a \) y \( b \)."

*   **Parámetros:**
    *   \( a \): Límite inferior del intervalo. $\( a < b \)$.
    *   \( b \): Límite superior del intervalo.

*   **Soporte:** $[a, b] $
    *   El conjunto de valores que la variable aleatoria puede tomar. Es un intervalo continuo.

*   **Función de Densidad de Probabilidad (FDP) - *Probability Density Function (PDF)*:**
    *   \( f(x) = \begin{cases} \frac{1}{b - a} & \text{para } a \le x \le b \\ 0 & \text{para otro caso} \end{cases} \)

*   **Función de Distribución Acumulada (FDA) - *Cumulative Distribution Function (CDF)*:**
    *   \( F(x) = \begin{cases} 0 & \text{para } x < a \\ \frac{x - a}{b - a} & \text{para } a \le x \le b \\ 1 & \text{para } x > b \end{cases} \)

*   **Esperanza (Media):**
    $$ E(X) = \frac{a + b}{2} $$

*   **Varianza :**
    $$ Var(X) = \frac{(b - a)^2}{12} $$

*   **Desviación Estándar :**
    $$ \sigma = \sqrt{Var(X)} = \frac{b - a}{\sqrt{12}} $$

*   **Moda:**
    *   Cualquier valor en el intervalo \( [a, b] \).

*   **Mediana :**
    $$ \text{Mediana} = \frac{a + b}{2} $$

**Método seleccionado**

Para simular la distribución uniforme continua en un intervalo $[a,b]$, el método más sencillo y directo es la transformada inversa, ya que la función de distribución acumulativa (CDF) de la uniforme es fácil de invertir. Además, no requiere cálculos complejos ni generación de variables auxiliares (como en aceptación-rechazo o Box-Muller).

**Pseudocodigo**

1. Generar un numero aleatorio $U \sim$uniforme  $[0, 1]$
2. Aplicar la transformada inversa de la CDF de la uniforme continua:
$$X=a+(b-a)U$$

3. Retornar $X$ como la variable simulada



**Código**

```python
import numpy as np
import matplotlib.pyplot as plt

def simular_uniforme(a, b, n=1000):
    return np.random.uniform(a, b, n)

# Parámetros y simulación
a=0
b=1
muestras = simular_uniforme(a, b)

#Valores
print('La varianza es:',np.var(muestras))
print('La media es:',np.mean(muestras))
print('La varianza teorica es:', (b-a)**2/12)
print('La media teorica es:', (a+b)/2)


# Configurar gráficos
plt.figure(figsize=(12, 4))

# Histograma (PDF)
plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='teal', edgecolor='black')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('PDF de Uniforme({}, {})'.format(a, b))

# Gráfico
plt.subplot(1, 2, 2)
plt.scatter(np.sort(muestras), np.linspace(0, 1, len(muestras)), color='green', s=10)
plt.plot([a, b], [0, 1], 'r--')
plt.title('Gráfico de la CDF')
plt.xlabel('Valor')
plt.ylabel('Probabilidad Acumulada')
plt.tight_layout()
plt.show()
```

##Triangular


**Nomenclatura**



*   **Notación:** $ X \sim \text{Triangular}(a, c, b)$
    *   "La variable aleatoria X sigue una distribución triangular continua con parámetros \( a \), \( c \) y \( b \)."

*   **Parámetros:**
    *   \( a \): Límite inferior o valor mínimo.
    *   \( c \): Valor modal o moda (la cumbre del triángulo). $\( a \le c \le b \)$.
    *   \( b \): Límite superior o valor máximo. $\( a < b \)$.

*   **Soporte:** $ [a, b] $
    *   El conjunto de valores que la variable aleatoria puede tomar.

*   **Función de Densidad de Probabilidad (FDP) - *Probability Density Function (PDF)*:**
    *   \( f(x) = \begin{cases} 0 & \text{para } x < a \\ \frac{2(x-a)}{(b-a)(c-a)} & \text{para } a \le x < c \\ \frac{2}{b-a} & \text{para } x = c \\ \frac{2(b-x)}{(b-a)(b-c)} & \text{para } c < x \le b \\ 0 & \text{para } x > b \end{cases} \)
    *   Representa la altura del triángulo en el punto \( x \).

*   **Función de Distribución Acumulada (FDA) - *Cumulative Distribution Function (CDF)*:**
    *   \( F(x) = \begin{cases} 0 & \text{para } x < a \\ \frac{(x-a)^2}{(b-a)(c-a)} & \text{para } a \le x < c \\ 1 - \frac{(b-x)^2}{(b-a)(b-c)} & \text{para } c \le x \le b \\ 1 & \text{para } x > b \end{cases} \)
    *   Representa la probabilidad de que la variable aleatoria sea menor o igual a un valor dado \( x \).

*   **Esperanza (Media) :**
    $$ E(X) = \frac{a + c + b}{3} $$

*   **Varianza:**
    $$ Var(X) = \frac{a^2 + b^2 + c^2 - ab - ac - bc}{18} $$

*   **Moda :**
    $$ \text{Moda} = c $$
    *   El valor con la mayor densidad de probabilidad.

*   **Mediana :**
    *   Si $\( c \ge \frac{a+b}{2} \)$, $$ \text{Mediana} = a + \sqrt{(b-a)(c-a)/2} $$
    *   Si $\( c \le \frac{a+b}{2} \)$, $$ \text{Mediana} = b - \sqrt{(b-a)(b-c)/2} $$
    *   El valor que divide el área del triángulo en dos partes iguales.

**Método seleccionado**

Para simular la distribución triangular continua (definida por los parámetros a: mínimo, b: máximo, y c: moda), el método más sencillo es la transformada inversa, ya que su CDF es invertible analíticamente.

**Pseudocodigo**

1. Generar $U \sim$ Uniforme$(0,1)$
2. Calcular la condicion para elegir el segmento de la CDF:


*   Si $U \leq \frac{c-a}{b-a}$:
$$X=a+\sqrt{U(b-a)(c-a)}$$
*   Si $U > \frac{c-a}{b-a}$:
$$X=b-\sqrt{(1-U)(b-a)(b-c)}$$

3. Retornar $X$


**Código**

```python
import numpy as np
import matplotlib.pyplot as plt

def simular_triangular(a, b, c, n=10000):
    U = np.random.rand(n)
    cond = U <= (b-a)/(c-a)
    X = np.where(cond,
                a + np.sqrt(U*(c-a)*(b-a)),
                c - np.sqrt((1-U)*(c-a)*(c-b)))
    return X

# Parámetros (a=mínimo, b=moda, c=máximo)
a=1
b=3
c=5
muestras = simular_triangular(a, b, c)

# Cálculos estadísticos
print('Media muestral:', np.mean(muestras))
print('Media teórica:', (a + b + c)/3)
print('\nVarianza muestral:', np.var(muestras))
print('Varianza teórica:', (a**2 + b**2 + c**2 - a*b - a*c - b*c)/18)

# Gráficos
plt.figure(figsize=(12, 4))

# PDF
plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='teal', edgecolor='black')
plt.title(f'PDF Triangular(a={a}, b={b}, c={c})')
plt.xlabel('Valor')
plt.ylabel('Densidad')

# CDF empírica
plt.subplot(1, 2, 2)
x_values = np.linspace(a, c, 100)
cdf_teorica = np.piecewise(x_values,
                          [x_values < a,
                           (x_values >= a) & (x_values <= b),
                           (x_values > b) & (x_values <= c),
                           x_values > c],
                          [0,
                           lambda x: (x-a)**2/((c-a)*(b-a)),
                           lambda x: 1 - (c-x)**2/((c-a)*(c-b)),
                           1])

plt.scatter(np.sort(muestras), np.linspace(0, 1, len(muestras)),
           color='green', s=10, label='Empírica')
plt.plot(x_values, cdf_teorica, 'r--', label='Teórica')  # Línea correcta
plt.legend()
```

##Exponencial

**Nomenclatura**

*   **Notación:** $ X \sim \text{Exp}(\lambda) $
    *   "La variable aleatoria \( X \) sigue una distribución exponencial con parámetro de tasa $ \lambda $."

*   **Parámetro:**
    *   $\( \lambda \)$: Tasa de decaimiento $\( \lambda > 0 \)$
    *   Alternativamente: $\( \beta = 1/\lambda \) $(parámetro de escala)

*   **Soporte:** $ [0, \infty)$
    *   Valores reales no negativos

*   **Función de Densidad de Probabilidad (PDF):**
    $$ f(x) = \begin{cases}
    \lambda e^{-\lambda x} & \text{para } x \geq 0 \\
    0 & \text{para } x < 0
    \end{cases} $$

*   **Función de Distribución Acumulada (CDF):**
    $$ F(x) = \begin{cases}
    1 - e^{-\lambda x} & \text{para } x \geq 0 \\
    0 & \text{para } x < 0
    \end{cases} $$

*   **Esperanza (Media):**
    $$ E(X) = \frac{1}{\lambda} $$

*   **Varianza:**
    $$ \text{Var}(X) = \frac{1}{\lambda^2} $$

*   **Moda:**
    $$ \text{Moda} = 0 $$

**Gráfica**

image.png

image.png

**Método seleccionado**

La exponencial tiene una CDF invertible analíticamente, lo que hace que el método de **transformada inversa** sea el más eficiente y exacto. No requiere aproximaciones ni cálculos complejos, solo una uniforme y una operación logarítmica.

**Pseudocodigo**

1. Generar U ~ Uniforme(0, 1)  
2. Calcular X = -ln(1 - U) / λ  
3. Retornar X ~ Exponencial(λ)  

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt

def simular_exponencial(lambd, n=1000):
    U = np.random.rand(n)
    return -np.log(1 - U) / lambd

# Parámetros y simulación
lambd = 0.5  # Tasa de la exponencial
muestras = simular_exponencial(lambd, 10000)


# Cálculos estadísticos

print('La media muestral es:', np.mean(muestras))
print('La varianza muestral es:', np.var(muestras))

print('\nLa media teórica es:', 1/lambd)
print('La varianza teórica es:', 1/lambd**2)

# Gráficos

plt.figure(figsize=(12, 5))

# Histograma (PDF empírica)
plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='teal',
         edgecolor='black', alpha=0.7, label='Empírica')

# PDF teórica
x = np.linspace(0, 10/lambd, 1000)
pdf_teorica = lambd * np.exp(-lambd * x)
plt.plot(x, pdf_teorica, 'r-', lw=2, label='Teórica')
plt.title(f'PDF Exponencial(λ={lambd})')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()

# CDF empírica vs teórica
plt.subplot(1, 2, 2)
x = np.linspace(0, 10/lambd, 1000)
cdf_empirica = np.array([np.mean(muestras <= v) for v in x])
cdf_teorica = 1 - np.exp(-lambd * x)
plt.plot(x, cdf_empirica, 'b-', label='Empírica')
plt.plot(x, cdf_teorica, 'r--', label='Teórica')
plt.title('CDF Exponencial')
plt.xlabel('Valor')
plt.ylabel('Probabilidad Acumulada')
plt.legend()

plt.tight_layout()
plt.show()
```

##m-Erlang

**Nomenclatura**

*   **Notación:** $ X \sim \text{Erlang}(m, \lambda) $ o $ X \sim \text{Gamma}(m, \lambda) $ con$ m $ entero
    *   "La variable aleatoria X sigue una distribución Erlang con parámetros$ \( m \)$ y $\( \lambda \)$."

*   **Parámetros:**
    *   $\( m \)$: Parámetro de forma (entero positivo). Número de etapas exponenciales independientes.
    *   $\( \lambda \)$: Parámetro de tasa (positivo). Tasa de ocurrencia en cada etapa. $\( \lambda > 0 \)$.

*   **Soporte:** $ [0, \infty) $
    *   El conjunto de valores que la variable aleatoria puede tomar (valores no negativos).

*   **Función de Densidad de Probabilidad (FDP) - *Probability Density Function (PDF)*:**
    *   \( f(x) = \begin{cases}
    \frac{\lambda^m x^{m-1} e^{-\lambda x}}{(m-1)!} & \text{para } x \ge 0 \\
    0 & \text{para } x < 0
    \end{cases} \)
    *   Donde $\( (m-1)! \) $es el factorial de$ \( m-1 \)$.

*   **Función de Distribución Acumulada (FDA) - *Cumulative Distribution Function (CDF)*:**
    *   \( F(x) = \begin{cases}
    1 - e^{-\lambda x} \sum_{k=0}^{m-1} \frac{(\lambda x)^k}{k!} & \text{para } x \ge 0 \\
    0 & \text{para } x < 0
    \end{cases} \)
    *   Representa la probabilidad acumulada hasta el valor \( x \).

*   **Esperanza (Media):**
    $$ E(X) = \frac{m}{\lambda} $$

*   **Varianza:**
    $$ Var(X) = \frac{m}{\lambda^2} $$

*   **Moda:**
    $$ \text{Moda} = \begin{cases}
    \frac{m-1}{\lambda} & \text{para } m \ge 1 \\
    0 & \text{para } m = 1
    \end{cases} $$
    *   El valor con mayor densidad de probabilidad.

*   **Mediana:**
    *   No tiene fórmula cerrada simple. Se obtiene resolviendo:
    $$ 1 - e^{-\lambda x} \sum_{k=0}^{m-1} \frac{(\lambda x)^k}{k!} = 0.5 $$
    *   El valor que divide la distribución en dos partes de igual probabilidad.

*   **Propiedad Aditiva:**
    *   Si $\( X_i \sim \text{Erlang}(m_i, \lambda) \)$ son independientes, entonces:
    $$ \sum_{i=1}^n X_i \sim \text{Erlang}\left(\sum_{i=1}^n m_i, \lambda\right) $$

*   **Relación con la Exponencial:**
    *   Cuando \( m = 1 \), se reduce a una distribución exponencial con tasa$ \( \lambda \)$.

**Gráfica**

image.png

image.png

**Método seleccionado**

Usaremos el método de transformada inversa junto con la propiedad de que una variable m-Erlang es la suma de m variables exponenciales independientes con el mismo parámetro $\lambda$.

**Pseudocodigo**

1. Generar $m$ números aleatorios $U_1, U_2, \dots,U_m \sim$ Uniforme $(0,1)$

2. Convertirlos en exponenciales:
$$E_i=-In(U_i)/\lambda$$

3. Sumar todas las exponanciales:
$$X= \sum_{i=1}^m E_i$$

4. Retornar X

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def simular_erlang(m, lamda, n=1000):
    U = np.random.rand(n, m)
    E = -np.log(U)/lamda
    X = np.sum(E, axis=1)
    return X

# Parámetros y simulación
m = 3
lamda = 0.5
muestras = simular_erlang(m, lamda)

# Cálculos estadísticos
print('Media muestral:', np.mean(muestras))
print('Media teórica:', m/lamda)
print('\nVarianza muestral:', np.var(muestras))
print('Varianza teórica:', m/(lamda**2))

# Configurar gráficos
plt.figure(figsize=(12, 4))

# Histograma (PDF)
plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='teal', edgecolor='black')
x = np.linspace(0, 20, 1000)
plt.plot(x, stats.erlang.pdf(x, m, scale=1/lamda), 'r-', lw=2)
plt.title(f'PDF m-Erlang(m={m}, λ={lamda})')
plt.xlabel('Valor')
plt.ylabel('Densidad')

# Gráfico CDF empírica vs teórica
plt.subplot(1, 2, 2)
plt.scatter(np.sort(muestras), np.linspace(0, 1, len(muestras)),
            color='green', s=10, label='Empírica')
plt.plot(x, stats.erlang.cdf(x, m, scale=1/lamda), 'r--', label='Teórica')
plt.title('CDF m-Erlang')
plt.xlabel('Valor')
plt.ylabel('Probabilidad Acumulada')
plt.legend()

plt.tight_layout()
plt.show()
```

##Gamma

**Nomenclatura**

*   **Notación:**
$ X \sim \text{Gamma}(\alpha, \beta) $ o  $X \sim \Gamma(\alpha, \beta) $
    *   "La variable aleatoria X sigue una distribución Gamma con parámetros \(\alpha\) y \(\beta\)."

*   **Parámetros:**
    *   $\(\alpha\)$: Parámetro de forma (positivo). Controla la forma de la distribución. \(\alpha > 0\)
    *   $\(\beta\)$: Parámetro de escala (positivo). Controla la dispersión. \(\beta > 0\)
    *   (Nota alternativa: Algunos textos usan $\(\lambda = 1/\beta\)$ como parámetro de tasa)

*   **Soporte:** $[0, \infty) $
    *   El conjunto de valores que la variable aleatoria puede tomar (valores no negativos).

*   **Función de Densidad de Probabilidad (FDP) - *Probability Density Function (PDF)*:**
    *   \( f(x) = \begin{cases}
    \frac{x^{\alpha-1} e^{-x/\beta}}{\beta^\alpha \Gamma(\alpha)} & \text{para } x \ge 0 \\
    0 & \text{para } x < 0
    \end{cases} \)
    *   Donde \(\Gamma(\alpha)\) es la función Gamma.

*   **Función de Distribución Acumulada (FDA) - *Cumulative Distribution Function (CDF)*:**
    *   \( F(x) = \begin{cases}
    \frac{\gamma(\alpha, x/\beta)}{\Gamma(\alpha)} & \text{para } x \ge 0 \\
    0 & \text{para } x < 0
    \end{cases} \)
    *   Donde \(\gamma(\alpha, z)\) es la función Gamma incompleta inferior.

*   **Esperanza (Media):**
    $$ E(X) = \alpha \beta $$

*   **Varianza:**
    $$ Var(X) = \alpha \beta^2 $$

*   **Moda:**
    $$ \text{Moda} = \begin{cases}
    (\alpha-1)\beta & \text{para } \alpha \ge 1 \\
    0 & \text{para } \alpha < 1
    \end{cases} $$
    *   El valor con mayor densidad de probabilidad.

*   **Mediana:**
    *   No tiene fórmula cerrada simple. Se obtiene resolviendo:
    $$ F(x) = 0.5 $$

*   **Función Generadora de Momentos:**
    $$ M_X(t) = (1 - \beta t)^{-\alpha} \quad \text{para } t < 1/\beta $$

*   **Propiedades Especiales:**
    1. **Caso Exponencial:** Cuando $\(\alpha = 1\)$, se reduce a una distribución exponencial.
    2. **Caso Erlang:** Cuando $\(\alpha\)$ es entero, se convierte en distribución Erlang.
    3. **Caso Chi-cuadrado:** Cuando $\(\alpha = \nu/2\) $y$ \(\beta = 2\)$, se convierte en $\(\chi^2(\nu)\)$.
    4. **Propiedad Aditiva:** La suma de Gammas independientes con mismo $\(\beta\)$ es Gamma.

*   **Relación con Otras Distribuciones:**
    *   Es la distribución conjugada prior para la precisión (inverso de la varianza) en distribuciones normales.
    *   Modela tiempos de espera para múltiples eventos en procesos de Poisson.

**Gráfica**

image.png

**Método seleccionado**

El método de aceptación-rechazo es el más adecuado para la distribución gamma cuando el parámetro de forma α no es entero. Es más sencillo de implementar que la transformada inversa (que requiere resolver integrales complejas) y más flexible que el método de composición o el de 12 uniformes. Box-Muller es específico para distribuciones normales, por lo que no aplica aquí.

**Pseudocodigo**

1. Establecer $\alpha$ y $\beta$
2. $X=0$
3. Para i desde 1 hasta $\alpha$

   Generar $U\sim$Uniforme$(0,1)$
4. $E=-In(U)/\beta$
5. $X=X+E$
6. Retornar X


**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
from math import gamma

def simular_gamma(alpha, beta, n=1000):
    muestras = np.zeros(n)
    for i in range(n):
        X = 0
        for _ in range(alpha):  # Suma α exponenciales
            U = np.random.rand()
            E = -np.log(U)/beta
            X += E
        muestras[i] = X
    return muestras

# Parámetros y simulación
alpha, beta = 3, 0.5  # Ejemplo: α entero
muestras = simular_gamma(alpha, beta, 10000)


# Cálculos estadísticos

print('La media muestral es:', np.mean(muestras))
print('La varianza muestral es:', np.var(muestras))

print('\nLa media teórica es:', alpha/beta)
print('La varianza teórica es:', alpha/(beta**2))


# Gráficos

plt.figure(figsize=(12, 5))

# Histograma (PDF empírica)
plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='teal',
         edgecolor='black', alpha=0.7, label='Empírica')

# PDF teórica
x = np.linspace(0, 20, 1000)
pdf_teorica = (beta**alpha) * (x**(alpha-1)) * np.exp(-beta*x) / gamma(alpha)
plt.plot(x, pdf_teorica, 'r-', lw=2, label='Teórica')
plt.title(f'PDF Gamma(α={alpha}, β={beta})')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()

# CDF empírica vs teórica
plt.subplot(1, 2, 2)
x = np.linspace(0, 20, 1000)
cdf_empirica = np.array([np.mean(muestras <= v) for v in x])
cdf_teorica = np.cumsum(pdf_teorica)*(x[1]-x[0])  # Aproximación numérica
plt.plot(x, cdf_empirica, 'b-', label='Empírica')
plt.plot(x, cdf_teorica, 'r--', label='Teórica')
plt.title('CDF Gamma')
plt.xlabel('Valor')
plt.ylabel('Probabilidad Acumulada')
plt.legend()

plt.tight_layout()
plt.show()
```

##Weibull

**Nomenclatura**

*   **Notación:** $ X \sim \text{Weibull}(k, \lambda) $
    *   "La variable aleatoria X sigue una distribución Weibull con parámetros de forma $\( k \)$ y escala $\( \lambda \)$."

*   **Parámetros:**
    *   $\( k \)$: Parámetro de forma (positivo). Controla el comportamiento de la tasa de fallo. $\( k > 0 \)$
    *   $\( \lambda \)$: Parámetro de escala (positivo).$ \( \lambda > 0 \)$
    *   (Alternativamente: Algunos textos usan $\( \alpha = \lambda^k \))$

*   **Soporte:** $[0, \infty)$
    *   Valores no negativos (modela tiempos de vida/duración).

*   **Función de Densidad de Probabilidad (FDP):**
    *   \( f(x) = \begin{cases}
    \frac{k}{\lambda} \left( \frac{x}{\lambda} \right)^{k-1} e^{-(x/\lambda)^k} & \text{para } x \geq 0 \\
    0 & \text{para } x < 0
    \end{cases} \)

*   **Función de Distribución Acumulada (FDA):**
    *   \( F(x) = \begin{cases}
    1 - e^{-(x/\lambda)^k} & \text{para } x \geq 0 \\
    0 & \text{para } x < 0
    \end{cases} \)

*   **Función de Supervivencia:**
    $$ S(x) = e^{-(x/\lambda)^k} \quad \text{para } x \geq 0 $$

*   **Función de Riesgo (Tasa de Fallo):**
    $$ h(x) = \frac{k}{\lambda} \left( \frac{x}{\lambda} \right)^{k-1} $$

*   **Esperanza (Media):**
    $$ E(X) = \lambda \, \Gamma\left(1 + \frac{1}{k}\right) $$

*   **Varianza:**
    $$ Var(X) = \lambda^2 \left[ \Gamma\left(1 + \frac{2}{k}\right) - \left( \Gamma\left(1 + \frac{1}{k}\right) \right)^2 \right] $$

*   **Moda:**
    $$ \text{Moda} = \begin{cases}
    \lambda \left( \frac{k-1}{k} \right)^{1/k} & \text{para } k \geq 1 \\
    0 & \text{para } 0 < k < 1
    \end{cases} $$

*   **Mediana:**
    $$ \text{Mediana} = \lambda (\ln 2)^{1/k} $$

*   **Casos Especiales:**
    1. **Exponencial:** Cuando \( k = 1 \), se reduce a distribución exponencial.
    2. **Rayleigh:** Cuando \( k = 2 \), se convierte en distribución Rayleigh.
    3. **Determinística:** Cuando $\( k \to \infty \)$, converge a $\( \lambda \)$.

*   **Aplicaciones Típicas:**
    *   Análisis de supervivencia y fiabilidad
    *   Modelado de tiempos de fallo en ingeniería
    *   Datos meteorológicos (distribución de vientos)

**Método seleccionado**

La distribución Weibull tiene una función de distribución acumulativa (CDF) invertible analíticamente, lo que hace que el método de transformada inversa sea el más simple y eficiente para generarla.

**Pseudocodigo**

1. Generar un número aleatorio $U$ ~ Uniforme $[0, 1]$
2. Aplicar la transformada inversa de la CDF de Weibull:
   $$X = λ * (-ln(1 - U))^(1/k)$$
3. Retornar $X$

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from math import gamma  # Importamos gamma desde math

def simular_weibull(lamda, k, n=1000):
    U = np.random.rand(n)
    X = lamda * (-np.log(1 - U))**(1/k)
    return X

# Parámetros y simulación
lamda = 2.0  # parámetro de escala
k = 1.5      # parámetro de forma
muestras = simular_weibull(lamda, k)

# Cálculos estadísticos
print('Media muestral:', np.mean(muestras))
print('Media teórica:', lamda * gamma(1 + 1/k))  # Usamos gamma de math
print('\nVarianza muestral:', np.var(muestras))
print('Varianza teórica:', lamda**2 * (gamma(1 + 2/k) - (gamma(1 + 1/k))**2))

# Configurar gráficos
plt.figure(figsize=(12, 4))

# Histograma (PDF)
plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='teal', edgecolor='black')
x = np.linspace(0, 10, 1000)
plt.plot(x, stats.weibull_min.pdf(x, k, scale=lamda), 'r-', lw=2)
plt.title(f'PDF Weibull(λ={lamda}, k={k})')
plt.xlabel('Valor')
plt.ylabel('Densidad')

# Gráfico CDF empírica vs teórica
plt.subplot(1, 2, 2)
plt.scatter(np.sort(muestras), np.linspace(0, 1, len(muestras)),
            color='green', s=10, label='Empírica')
plt.plot(x, stats.weibull_min.cdf(x, k, scale=lamda), 'r--', label='Teórica')
plt.title('CDF Weibull')
plt.xlabel('Valor')
plt.ylabel('Probabilidad Acumulada')
plt.legend()

plt.tight_layout()
plt.show()
```

##Normal



**Nomenclatura**

*   **Notación:** $ X \sim N(\mu, \sigma^2) $
    *   "La variable aleatoria X sigue una distribución normal con media $\(\mu\)$ y varianza $\(\sigma^2\)$."

*   **Parámetros:**
    *   $\(\mu\)$: Parámetro de ubicación (media). Puede ser cualquier valor real.
    *   $\(\sigma\)$: Parámetro de escala (desviación estándar). \(\sigma > 0\).

*   **Soporte:** $ (-\infty, \infty) $
    *   Todos los números reales.

*   **Función de Densidad de Probabilidad (FDP):**
    $$ f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2} $$

*   **Función de Distribución Acumulada (FDA):**
    *   No tiene forma cerrada. Se expresa mediante:
    $$ F(x) = \Phi\left(\frac{x-\mu}{\sigma}\right) $$
    *   Donde $\(\Phi\) $es la FDA de la normal estándar $\(N(0,1)\)$.

*   **Esperanza (Media):**
    $$ E(X) = \mu $$

*   **Varianza:**
    $$ Var(X) = \sigma^2 $$

*   **Desviación Estándar:**
    $$ \sigma = \sqrt{Var(X)} $$

*   **Moda:**
    $$ \text{Moda} = \mu $$
    *   Coincide con la media y mediana.

*   **Mediana:**
    $$ \text{Mediana} = \mu $$

*   **Simetría:**
    $$ \text{Sesgo} = 0 $$
    *   La distribución es perfectamente simétrica.

*   **Curtosis:**
    $$ \text{Curtosis} = 3 $$
    *   La curtosis excesiva es 0 (comparada con la normal).

*   **Estandarización:**
    $$ Z = \frac{X - \mu}{\sigma} \sim N(0, 1) $$

*   **Propiedad Aditiva:**
    *   Si $\(X_i \sim N(\mu_i, \sigma_i^2)\)$ son independientes:
    $$ \sum_{i=1}^n X_i \sim N\left(\sum_{i=1}^n \mu_i, \sum_{i=1}^n \sigma_i^2\right) $$

*   **Teorema del Límite Central:**
    *   La suma de $\(n\) $variables aleatorias i.i.d. converge a $\(N(n\mu, n\sigma^2)\) $cuando $\(n \to \infty\)$.

*   **Aplicaciones:**
    1. Modelado de errores de medición
    2. Fenómenos naturales (alturas, pesos)
    3. Inferencia estadística
    4. Finanzas (movimientos de precios)

**Método seleccionado**

Seleccionamos el metodo de Box Muller ya que es el método más sencillo y eficiente para generar variables normales exactas (no aproximadas) a partir de números aleatorios uniformes

**Pseudocodigo**

1. Generar $U_1, U_2 \sim $ Uniforme $[0, 1]$ independientes.  
2. Calcular:  
   $$Z₀ = √(-2·ln(U_1)) · cos(2πU_2)  $$
   $$Z₁ = √(-2·ln(U_1)) · sin(2πU_2)  $$
3. Retornar $(Z_0, Z_1) $

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, sqrt, exp
from scipy.special import erf  # Importación añadida para la función error

def box_muller(n=1000):
    """
    Simula n muestras de una distribución Normal(0,1) usando Box-Muller
    Parámetros:
        n: número de muestras (debe ser par)
    Devuelve:
        np.array: muestras Normales estándar
    """
    if n % 2 != 0:
        n += 1  # Asegurar n par

    U1 = np.random.rand(n//2)
    U2 = np.random.rand(n//2)

    Z0 = np.sqrt(-2 * np.log(U1)) * np.cos(2 * pi * U2)
    Z1 = np.sqrt(-2 * np.log(U1)) * np.sin(2 * pi * U2)

    return np.concatenate([Z0, Z1])

# Parámetros y simulación
muestras = box_muller(10000)  # 10000 muestras N(0,1)


# Cálculos estadísticos

print('La media muestral es:', np.mean(muestras))
print('La varianza muestral es:', np.var(muestras))

print('\nLa media teórica es:', 0)
print('La varianza teórica es:', 1)


# Gráficos

plt.figure(figsize=(12, 5))

# Histograma (PDF empírica)
plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='teal', edgecolor='black', alpha=0.7, label='Empírica')

# PDF teórica
x = np.linspace(-4, 4, 1000)
pdf_teorica = np.exp(-x**2/2) / sqrt(2*pi)
plt.plot(x, pdf_teorica, 'r-', lw=2, label='Teórica')
plt.title('PDF Normal Estándar')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()

# CDF empírica vs teórica
plt.subplot(1, 2, 2)
cdf_empirica = np.array([np.mean(muestras <= v) for v in x])
cdf_teorica = (1 + erf(x/sqrt(2))) / 2  # Usando la función error importada
plt.plot(x, cdf_empirica, 'b-', label='Empírica')
plt.plot(x, cdf_teorica, 'r--', label='Teórica')
plt.title('CDF Normal Estándar')
plt.xlabel('Valor')
plt.ylabel('Probabilidad Acumulada')
plt.legend()

plt.tight_layout()
plt.show()
```

##Log-normal

**Nomenclatura**

*   **Notación:** $ X \sim \text{LogN}(\mu, \sigma^2) $
    *   "La variable aleatoria X sigue una distribución log-normal con parámetros $\(\mu\) $ y $\(\sigma^2\)$."

*   **Relación con la Normal:**
    *   $ Y = \ln(X) \sim N(\mu, \sigma^2) $
    *   $X = e^Y $ donde $ Y \sim N(\mu, \sigma^2)$

*   **Parámetros:**
    *   $\(\mu\)$: Media del logaritmo de la variable (parámetro de ubicación). $\(\mu \in \mathbb{R}\)$
    *   $\(\sigma\)$: Desviación estándar del logaritmo (parámetro de escala).$ \(\sigma > 0\)$

*   **Soporte:** $(0, \infty) $
    *   Valores estrictamente positivos.

*   **Función de Densidad de Probabilidad (FDP):**
    $$ f(x) = \frac{1}{x\sigma\sqrt{2\pi}} e^{-\frac{(\ln x - \mu)^2}{2\sigma^2}} \quad \text{para } x > 0 $$

*   **Función de Distribución Acumulada (FDA):**
    $$ F(x) = \Phi\left(\frac{\ln x - \mu}{\sigma}\right) $$
    *   Donde \(\Phi\) es la FDA de la normal estándar.

*   **Esperanza (Media):**
    $$ E(X) = e^{\mu + \frac{\sigma^2}{2}} $$

*   **Varianza:**
    $$ Var(X) = \left(e^{\sigma^2} - 1\right) e^{2\mu + \sigma^2} $$

*   **Moda:**
    $$ \text{Moda} = e^{\mu - \sigma^2} $$

*   **Mediana:**
    $$ \text{Mediana} = e^\mu $$

*   **Asimetría:**
    $$ \text{Sesgo} = \left(e^{\sigma^2} + 2\right)\sqrt{e^{\sigma^2} - 1} $$
    *   Siempre positiva (cola derecha más larga).

*   **Curtosis:**
    $$ \text{Curtosis} = e^{4\sigma^2} + 2e^{3\sigma^2} + 3e^{2\sigma^2} - 6 $$

image.png

**Método seleccionado**

La transformación de normales (vía Box-Muller + exponencial) es el método más sencillo, exacto y eficiente para generar variables log-normales.

**Pseudocodigo**

1. Generar $Z \sim $ Normal$(0,1) $usando Box-Muller:  
   1.1. Generar $U_1, U_2 \sim $ Uniforme$[0,1]$  
   1.2. Calcular:

   $$Z = \sqrt{(-2 ln U_1) · cos(2π U_2)}$$  

   
2. Aplicar transformación exponencial:  
   $$X = e^{(μ + σ Z)}$$
3. Retornar $X$

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, sqrt, exp
from scipy.special import erf  # Importar erf desde scipy.special

def box_muller(n=1000):
    m = (n + 1) // 2  # Para asegurar suficientes muestras

    U1 = np.random.rand(m)
    U2 = np.random.rand(m)

    Z0 = np.sqrt(-2 * np.log(U1)) * np.cos(2 * pi * U2)
    Z1 = np.sqrt(-2 * np.log(U1)) * np.sin(2 * pi * U2)

    return np.concatenate([Z0, Z1])[:n]

def log_normal(mu, sigma, n=1000):
    Z = box_muller(n)
    X = np.exp(mu + sigma * Z)
    return X

# Parámetros y simulación
mu = 0      # Parámetro mu de la log-normal
sigma = 1   # Parámetro sigma de la log-normal
muestras = log_normal(mu, sigma, 10000)  # 10,000 muestras log-normales


# Cálculos estadísticos

# Momentos teóricos de la log-normal
teorica_media = exp(mu + sigma**2/2)
teorica_varianza = (exp(sigma**2) - 1) * exp(2*mu + sigma**2)

print('La media muestral es:', np.mean(muestras))
print('La varianza muestral es:', np.var(muestras))

print('\nLa media teórica es:', teorica_media)
print('La varianza teórica es:', teorica_varianza)


# Gráficos

plt.figure(figsize=(12, 5))

# Histograma (PDF empírica)
plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='skyblue',
         edgecolor='black', alpha=0.7, label='Empírica', range=(0, 10))

# PDF teórica
x = np.linspace(0.01, 10, 1000)
pdf_teorica = (1 / (x * sigma * sqrt(2*pi))) * np.exp(-(np.log(x)-mu)**2 / (2*sigma**2))
plt.plot(x, pdf_teorica, 'r-', lw=2, label='Teórica')
plt.title('PDF Log-Normal(μ={}, σ={})'.format(mu, sigma))
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()

# CDF empírica vs teórica
plt.subplot(1, 2, 2)
cdf_empirica = np.array([np.mean(muestras <= v) for v in x])
cdf_teorica = 0.5 + 0.5 * erf((np.log(x)-mu)/(sigma*sqrt(2)))  # Usando erf de scipy
plt.plot(x, cdf_empirica, 'b-', label='Empírica')
plt.plot(x, cdf_teorica, 'r--', label='Teórica')
plt.title('CDF Log-Normal(μ={}, σ={})'.format(mu, sigma))
plt.xlabel('Valor')
plt.ylabel('Probabilidad Acumulada')
plt.legend()

plt.tight_layout()
plt.show()
```

##Beta

**Nomenclatura**

*   **Notación:** $ X \sim \text{Beta}(\alpha, \beta) $
    *   "La variable aleatoria \( X \) sigue una distribución beta con parámetros de forma $\( \alpha \)$ y $\( \beta \)$."

*   **Parámetros:**
    *   $\( \alpha \)$: Parámetro de forma 1 $(\alpha > 0\)$
    *   $\( \beta \)$: Parámetro de forma 2 $(\beta > 0\)$

*   **Soporte:** $[0, 1] $
    *   Valores en el intervalo unitario (puede extenderse a cualquier intervalo finito)

*   **Función de Densidad de Probabilidad (PDF):**
    $$ f(x) = \begin{cases}
    \frac{x^{\alpha-1}(1-x)^{\beta-1}}{B(\alpha, \beta)} & \text{para } 0 \leq x \leq 1 \\
    0 & \text{otro caso}
    \end{cases} $$
    *   Donde $ B(\alpha, \beta) = \frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha+\beta)} $ es la función beta

*   **Función de Distribución Acumulada (CDF):**
    $$ F(x) = I_x(\alpha, \beta) $$
    *   Donde $ I_x $ es la función beta incompleta regularizada

*   **Esperanza (Media):**
    $$ E(X) = \frac{\alpha}{\alpha + \beta} $$

*   **Varianza:**
    $$ Var(X) = \frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)} $$

*   **Moda:**
    $$ \text{Moda} = \begin{cases}
    \frac{\alpha-1}{\alpha+\beta-2} & \text{para } \alpha, \beta > 1 \\
    0 \text{ o } 1 & \text{para } \alpha, \beta < 1 \\
    \text{indefinida} & \text{para } \alpha = \beta = 1 \text{ (uniforme)}
    \end{cases} $$

**Método seleccionado**

Método de Composición
La distribución Beta está relacionada con la razón de variables Gamma. Si $Y_1 \sim$ Gamma $(\alpha, 1)$ y $Y_2 \sim$ Gamma $(b, 1)$, entonces:
$$X= \frac{Y_1}{Y_1+Y_2}\sim Beta(a,b)$$

Este método es exacto, eficiente y fácil de implementar si ya se tiene un generador de Gamma (como el de suma de exponenciales para parámetros enteros o aceptación-rechazo para no enteros).

**Pseudocodigo**

1. Calcular constante $M = \max f(x)$ donde:
   $$f(x) = \frac{x^{\alpha-1}(1-x)^{\beta-1}}{B(\alpha,\beta)}$$

2. Generar muestras:

   2.1. Generar $X \sim \text{Uniforme}(0,1)$

   2.2. Generar $U \sim \text{Uniforme}(0,1)$
   
3. Aceptar $X$ si:
   $$U \leq \frac{f(X)}{M}$$

4. Repetir hasta obtener $n$ muestras aceptadas

5. Retornar muestras aceptadas



**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
from math import gamma
from scipy.stats import beta as beta_dist

def beta_pdf(x, alpha, beta):
    """Función de densidad de Beta(alpha, beta)."""
    return (x**(alpha-1) * (1-x)**(beta-1)) / (gamma(alpha) * gamma(beta) / gamma(alpha + beta))

def simular_beta_ar(alpha, beta, n=1000):

    muestras = []
    M = beta_pdf((alpha-1)/(alpha+beta-2), alpha, beta)  # Máximo de f(x)

    while len(muestras) < n:
        X = np.random.uniform(0, 1)  # Distribución propuesta (Uniforme)
        U = np.random.uniform(0, 1)  # Uniforme para la condición

        if U <= beta_pdf(X, alpha, beta) / M:
            muestras.append(X)

    return np.array(muestras)

# Parámetros y simulación
alpha, beta = 2, 5
muestras = simular_beta_ar(alpha, beta, 10000)  # 10,000 muestras Beta(2,5)


# Cálculos estadísticos

teorica_media = alpha / (alpha + beta)
teorica_varianza = (alpha * beta) / ((alpha + beta)**2 * (alpha + beta + 1))

print('Media muestral:', np.mean(muestras))
print('Varianza muestral:', np.var(muestras))
print('\nMedia teórica:', teorica_media)
print('Varianza teórica:', teorica_varianza)


# Gráficos

plt.figure(figsize=(12, 5))

# Histograma (PDF empírica)
plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='skyblue',
         edgecolor='black', alpha=0.7, label='Empírica', range=(0, 1))

# PDF teórica
x = np.linspace(0, 1, 1000)
pdf_teorica = beta_dist.pdf(x, alpha, beta)
plt.plot(x, pdf_teorica, 'r-', lw=2, label='Teórica')
plt.title(f'PDF Beta(α={alpha}, β={beta})')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()

# CDF empírica vs teórica
plt.subplot(1, 2, 2)
cdf_empirica = np.array([np.mean(muestras <= v) for v in x])
cdf_teorica = beta_dist.cdf(x, alpha, beta)
plt.plot(x, cdf_empirica, 'b-', label='Empírica')
plt.plot(x, cdf_teorica, 'r--', label='Teórica')
plt.title(f'CDF Beta(α={alpha}, β={beta})')
plt.xlabel('Valor')
plt.ylabel('Probabilidad Acumulada')
plt.legend()

plt.tight_layout()
plt.show()
```

##Pearson tipos

###Tipo VII: Distribución t-Student

**Nomenclatura**

*   **Notación:** $ X \sim t_\nu $
    *   "La variable aleatoria $\( X \)$ sigue una distribución t-Student con \$( \nu \)$ grados de libertad."

*   **Parámetro:**
    *   $\( \nu \)$: Grados de libertad $( \nu > 0 \)$, típicamente entero positivo

*   **Soporte:** $ (-\infty, \infty) $
    *   Todos los números reales

*   **Función de Densidad de Probabilidad (PDF):**
    $$ f(x) = \frac{\Gamma\left(\frac{\nu+1}{2}\right)}{\sqrt{\nu\pi}\,\Gamma\left(\frac{\nu}{2}\right)}\left(1+\frac{x^2}{\nu}\right)^{-\frac{\nu+1}{2}} $$

*   **Función de Distribución Acumulada (CDF):**
    $$ F(x) = \frac{1}{2} + x \Gamma\left(\frac{\nu+1}{2}\right) \cdot \frac{{}_2F_1\left(\frac{1}{2}, \frac{\nu+1}{2}; \frac{3}{2}; -\frac{x^2}{\nu}\right)}{\sqrt{\nu\pi}\,\Gamma\left(\frac{\nu}{2}\right)} $$
    *   Donde $\( {}_2F_1 \)$ es la función hipergeométrica

*   **Esperanza (Media):**
    $$ E(X) = \begin{cases}
    0 & \text{para } \nu > 1 \\
    \text{indefinida} & \text{para } \nu \leq 1
    \end{cases} $$

*   **Varianza:**
    $$ \text{Var}(X) = \begin{cases}
    \frac{\nu}{\nu-2} & \text{para } \nu > 2 \\
    \infty & \text{para } 1 < \nu \leq 2 \\
    \text{indefinida} & \text{para } \nu \leq 1
    \end{cases} $$

*   **Moda:**
    $$ \text{Moda} = 0 $$

*   **Asimetría:**
    $$ \text{Sesgo} = 0 \quad \text{para } \nu > 3 $$

*   **Curtosis:**
    $$ \text{Curtosis} = \begin{cases}
    \frac{6}{\nu-4} & \text{para } \nu > 4 \\
    \infty & \text{para } 2 < \nu \leq 4 \\
    \text{indefinida} & \text{para } \nu \leq 2
    \end{cases} $$

**Método seleccionado**

Box-Muller es ideal para simular la t-Student porque genera normales exactas y eficientemente, que son la base de su definición. Evita aproximaciones y es más rápido que otros métodos como aceptación-rechazo o transformada inversa.

**Pseudocodigo**

1. Generar $Z \sim \text{Normal}(0,1)$ usando Box-Muller:

   1.1. Generar $U_1, U_2 \sim \text{Uniforme}(0,1)$

   1.2. Calcular $Z = \sqrt{-2\ln U_1} \cdot \cos(2\pi U_2)$

2. Generar $V \sim \sqrt{\chi^2_\nu / \nu}$:

   2.1. Generar $Y \sim \text{Gamma}(\nu/2, 2)$ (equivale a $\chi^2_\nu$)

   2.2. Calcular $V = \sqrt{Y / \nu}$

3. Calcular:

   $$T = \frac{Z}{V} \sim t_\nu$$

4. Retornar $T$

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

def box_muller(n):
    m = (n + 1) // 2  # Asegurar suficientes muestras
    U1 = np.random.rand(m)
    U2 = np.random.rand(m)
    Z0 = np.sqrt(-2 * np.log(U1)) * np.cos(2 * np.pi * U2)
    Z1 = np.sqrt(-2 * np.log(U1)) * np.sin(2 * np.pi * U2)
    return np.concatenate([Z0, Z1])[:n]  # Retorna exactamente n muestras

def simular_tstudent(nu, n=1000):
    Z = box_muller(n)  # Normal(0,1) via Box-Muller (n muestras exactas)
    Y = np.random.gamma(shape=nu/2, scale=2, size=n)  # Chi-cuadrado(nu) = Gamma(nu/2, 2)
    V = np.sqrt(Y / nu)  # Raíz de chi-cuadrado/nu
    T = Z / V  # t-Student
    return T

# Parámetros y simulación
nu = 5  # Grados de libertad
muestras = simular_tstudent(nu, 10000)


# Cálculos estadísticos

print('Media muestral:', np.mean(muestras))
print('Varianza muestral:', np.var(muestras))
print('\nMedia teórica:', 0 if nu > 1 else 'Indefinida')
print('Varianza teórica:', nu / (nu - 2) if nu > 2 else 'Infinito')

# Gráficos

plt.figure(figsize=(12, 5))

# Histograma (PDF empírica)
plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='teal',
         edgecolor='black', alpha=0.7, label='Empírica', range=(-5, 5))

# PDF teórica
x = np.linspace(-5, 5, 1000)
pdf_teorica = t.pdf(x, nu)
plt.plot(x, pdf_teorica, 'r-', lw=2, label='Teórica')
plt.title(f'PDF t-Student ($\\nu={nu}$)')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()

# CDF empírica vs teórica
plt.subplot(1, 2, 2)
cdf_empirica = np.array([np.mean(muestras <= v) for v in x])
cdf_teorica = t.cdf(x, nu)
plt.plot(x, cdf_empirica, 'b-', label='Empírica')
plt.plot(x, cdf_teorica, 'r--', label='Teórica')
plt.title(f'CDF t-Student ($\\nu={nu}$)')
plt.xlabel('Valor')
plt.ylabel('Probabilidad Acumulada')
plt.legend()

plt.tight_layout()
plt.show()
```

### Tipo VI: Distribución F (beta prima)

**Nomenclatura**

*   **Notación:** $ X \sim \text{BetaPrime}(\alpha, \beta)$
    *   "La variable aleatoria $\( X \) $sigue una distribución Beta Prime con parámetros$ \( \alpha \) $y $\( \beta \)$."
    *   También conocida como Beta de Primer Tipo o Pearson Tipo VI

*   **Parámetros:**
    *   $\( \alpha \)$: Parámetro de forma $\( \alpha > 0 \)$
    *   $\( \beta \)$: Parámetro de forma $\( \beta > 0 \)$
    *   Opcionalmente: $\( \lambda \)$ parámetro de escala $\( \lambda > 0 \)$

*   **Soporte:** $ (0, \infty) $
    *   Valores reales positivos

*   **Función de Densidad de Probabilidad (PDF):**
    $$ f(x) = \frac{x^{\alpha-1} (1+x)^{-\alpha-\beta}}{B(\alpha, \beta)} \quad \text{para } x > 0 $$
    *   Donde \( B(\alpha, \beta) \) es la función Beta

*   **Función de Distribución Acumulada (CDF):**
    $$ F(x) = I_{\frac{x}{1+x}}(\alpha, \beta) $$
    *   Donde \( I \) es la función Beta incompleta regularizada

*   **Esperanza (Media):**
    $$ E(X) = \begin{cases}
    \frac{\alpha}{\beta-1} & \text{para } \beta > 1 \\
    \text{indefinida} & \text{para } \beta \leq 1
    \end{cases} $$

*   **Varianza:**
    $$ \text{Var}(X) = \begin{cases}
    \frac{\alpha(\alpha+\beta-1)}{(\beta-1)^2(\beta-2)} & \text{para } \beta > 2 \\
    \text{indefinida} & \text{para } \beta \leq 2
    \end{cases} $$

*   **Moda:**
    $$ \text{Moda} = \begin{cases}
    \frac{\alpha-1}{\beta+1} & \text{para } \alpha \geq 1 \\
    0 & \text{para } \alpha < 1
    \end{cases} $$

image.png

**Método seleccionado**

 La mejor opción para simular una Pearson Tipo VI (Beta prima) es aceptación-rechazo, ya que no hay una transformación directa con las otras opciones.

**Pseudocodigo**

1. \textbf{Definir} función de densidad objetivo:
   $$f(x) = \frac{x^{\alpha-1} (1+x)^{-\alpha-\beta}}{B(\alpha,\beta)}$$

2. \textbf{Elegir distribución propuesta} $g(x)$ y constante $M$:
   - Usar $g(x) = \text{Gamma}(\alpha, 1)$ (similar para cola derecha)
   - Calcular $M = \max \left(\frac{f(x)}{g(x)}\right)$

3. \textbf{Generar muestras}:
   3.1. Generar $Y \sim \text{Gamma}(\alpha, 1)$
   3.2. Generar $U \sim \text{Uniforme}(0,1)$
   3.3. \textbf{Aceptar} $Y$ si:
   $$U \leq \frac{f(Y)}{M \cdot g(Y)}$$
   3.4. Repetir hasta obtener $n$ muestras

4. \textbf{Retornar} muestras aceptadas


**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import betaprime, gamma
from math import gamma as math_gamma

def beta_prime_pdf(x, alpha, beta):
    return (x**(alpha-1) * ((1 + x)**(-alpha - beta))) / (math_gamma(alpha) * math_gamma(beta) / math_gamma(alpha + beta))

def simular_betaprime_ar(alpha, beta, n=1000):
    muestras = []
    # Calcular M óptimo (puede ajustarse manualmente si es necesario)
    M = (math_gamma(beta) / (math_gamma(alpha + beta))) * (alpha + beta)**(alpha + beta) / (alpha**alpha * beta**beta)

    while len(muestras) < n:
        Y = np.random.gamma(shape=alpha, scale=1)  # Propuesta: Gamma(alpha, 1)
        U = np.random.uniform(0, 1)

        if U <= beta_prime_pdf(Y, alpha, beta) / (M * gamma.pdf(Y, alpha, scale=1)):
            muestras.append(Y)

    return np.array(muestras)

# Parámetros y simulación
alpha, beta = 2.5, 3.0
muestras = simular_betaprime_ar(alpha, beta, 10000)  # 10,000 muestras


# Cálculos estadísticos

teorica_media = alpha / (beta - 1) if beta > 1 else np.nan
teorica_varianza = alpha * (alpha + beta - 1) / ((beta - 2) * (beta - 1)**2) if beta > 2 else np.nan

print('Media muestral:', np.mean(muestras))
print('Varianza muestral:', np.var(muestras))
print('\nMedia teórica:', teorica_media)
print('Varianza teórica:', teorica_varianza)

# Gráficos

plt.figure(figsize=(12, 5))

# Histograma (PDF empírica)
plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='skyblue',
         edgecolor='black', alpha=0.7, label='Empírica', range=(0, 10))

# PDF teórica
x = np.linspace(0, 10, 1000)
pdf_teorica = betaprime.pdf(x, alpha, beta)
plt.plot(x, pdf_teorica, 'r-', lw=2, label='Teórica')
plt.title(f'PDF Beta prima(α={alpha}, β={beta})')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()

# CDF empírica vs teórica
plt.subplot(1, 2, 2)
cdf_empirica = np.array([np.mean(muestras <= v) for v in x])
cdf_teorica = betaprime.cdf(x, alpha, beta)
plt.plot(x, cdf_empirica, 'b-', label='Empírica')
plt.plot(x, cdf_teorica, 'r--', label='Teórica')
plt.title(f'CDF Beta prima(α={alpha}, β={beta})')
plt.xlabel('Valor')
plt.ylabel('Probabilidad Acumulada')
plt.legend()

plt.tight_layout()
plt.show()
```

##Log-logística

**Nomenclatura**

*   **Notación:** $X \sim \text{LL}(\alpha, \beta) $
    *   "La variable aleatoria X sigue una distribución log-logística con parámetros de escala $\(\alpha\)$ y forma $\(\beta\)$."

*   **Relación con la Logística:**
    *   $Y = \ln(X) \sim \text{Logística}(\mu = \ln(\alpha), s = 1/\beta)$

*   **Parámetros:**
    *   $\(\alpha\)$: Parámetro de escala $(\alpha > 0\)$. Afecta la dispersión.
    *   $\(\beta\)$: Parámetro de forma $(\beta > 0\)$. Controla la forma de la distribución.

*   **Soporte:** $ (0, \infty)$
    *   Valores estrictamente positivos.

*   **Función de Densidad de Probabilidad (FDP):**
    $$ f(x) = \frac{(\beta/\alpha)(x/\alpha)^{\beta-1}}{\left[1 + (x/\alpha)^\beta\right]^2} \quad \text{para } x > 0 $$

*   **Función de Distribución Acumulada (FDA):**
    $$ F(x) = \frac{(x/\alpha)^\beta}{1 + (x/\alpha)^\beta} \quad \text{para } x > 0 $$

*   **Función de Supervivencia:**
    $$ S(x) = \frac{1}{1 + (x/\alpha)^\beta} $$

*   **Función de Riesgo:**
    $$ h(x) = \frac{(\beta/\alpha)(x/\alpha)^{\beta-1}}{1 + (x/\alpha)^\beta} $$

*   **Esperanza (Media):**
    $$ E(X) = \frac{\alpha \pi/\beta}{\sin(\pi/\beta)} \quad \text{para } \beta > 1 $$
    *   No existe para $\(\beta \leq 1\)$.

*   **Varianza:**
    $$ Var(X) = \alpha^2 \left[ \frac{2\pi/\beta}{\sin(2\pi/\beta)} - \frac{(\pi/\beta)^2}{\sin^2(\pi/\beta)} \right] \quad \text{para } \beta > 2 $$

*   **Moda:**
    $$ \text{Moda} = \alpha \left(\frac{\beta - 1}{\beta + 1}\right)^{1/\beta} \quad \text{para } \beta \geq 1 $$
    *   Si $\(\beta < 1\)$, la moda es 0.

*   **Mediana:**
    $$ \text{Mediana} = \alpha $$

*   **Asimetría:**
    *   Depende de $\(\beta\)$:
        - $\(\beta \leq 3\)$: Asimetría positiva
        - $\(\beta > 3\)$: Más simétrica

**Método seleccionado**

La distribución log-logística tiene una función de distribución acumulativa (CDF) invertible analíticamente, lo que hace que el método de transformada inversa sea el más directo y eficiente. No requiere aproximaciones ni algoritmos complejos.

**Pseudocodigo**

1. Generar $U ~$ Uniforme$(0, 1) $
2. Aplicar la transformada inversa de la CDF log-logística:  
   $$X = α * (U / (1 - U))^{(1/β)}  $$
3. Retornar $X $

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def simular_loglogistica(alpha, beta, n=1000):
    U = np.random.rand(n)
    X = alpha * (U / (1 - U))**(1/beta)
    return X

# Parámetros y simulación
alpha = 2.0  # parámetro de escala
beta = 3.0   # parámetro de forma
muestras = simular_loglogistica(alpha, beta)

# Cálculos estadísticos
mean_teorica = (alpha * np.pi / beta) / np.sin(np.pi / beta) if beta > 1 else np.inf
var_teorica = (alpha**2 * 2 * np.pi / beta) / np.sin(2 * np.pi / beta) - mean_teorica**2 if beta > 2 else np.inf

print('Media muestral:', np.mean(muestras))
print('Media teórica:', mean_teorica)
print('\nVarianza muestral:', np.var(muestras))
print('Varianza teórica:', var_teorica)

# Configurar gráficos
plt.figure(figsize=(12, 4))

# Histograma (PDF)
plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='teal', edgecolor='black', range=(0, 10))
x = np.linspace(0.01, 10, 1000)
pdf_teorica = (beta/alpha) * (x/alpha)**(beta-1) / (1 + (x/alpha)**beta)**2
plt.plot(x, pdf_teorica, 'r-', lw=2)
plt.title(f'PDF Log-Logística(α={alpha}, β={beta})')
plt.xlabel('Valor')
plt.ylabel('Densidad')

# Gráfico CDF empírica vs teórica
plt.subplot(1, 2, 2)
plt.scatter(np.sort(muestras), np.linspace(0, 1, len(muestras)),
            color='green', s=10, label='Empírica')
cdf_teorica = 1 / (1 + (x/alpha)**(-beta))
plt.plot(x, cdf_teorica, 'r--', label='Teórica')
plt.title('CDF Log-Logística')
plt.xlabel('Valor')
plt.ylabel('Probabilidad Acumulada')
plt.legend()

plt.tight_layout()
plt.show()
```

##Johnson


**Nomenclatura**

*   **Notación:** $X \sim \text{Johnson}(\gamma, \delta, \lambda, \xi)$
    *   "La variable aleatoria X sigue una distribución Johnson con parámetros $\gamma$, $\delta$, $\lambda$ y $\xi$."

*   **Tipos de Transformación:**
    1. **$S_U$ (No acotada):**
       $$ Z = \gamma + \delta \sinh^{-1}\left(\frac{X-\xi}{\lambda}\right) \sim N(0,1) $$
    2. **$S_B$ (Acotada):**
       $$ Z = \gamma + \delta \ln\left(\frac{X-\xi}{\xi+\lambda-X}\right) \sim N(0,1) $$

*   **Parámetros:**
    *   $\gamma$: Parámetro de forma (sesgo)
    *   $\delta$: Parámetro de forma (curtosis), $\delta > 0$
    *   $\lambda$: Parámetro de escala, $\lambda > 0$
    *   $\xi$: Parámetro de ubicación

*   **Soporte:**
    *   **$S_U$:** $(-\infty, \infty)$
    *   **$S_B$:** $(\xi, \xi+\lambda)$

*   **Función de Densidad (PDF):**
    *   **Para $S_U$:**
        $$ f(x) = \frac{\delta}{\lambda\sqrt{2\pi}} \frac{e^{-\frac{1}{2}\left(\gamma + \delta \sinh^{-1}\left(\frac{x-\xi}{\lambda}\right)\right)^2}}{\sqrt{1 + \left(\frac{x-\xi}{\lambda}\right)^2}} $$
    *   **Para $S_B$:**
        $$ f(x) = \frac{\delta}{\lambda\sqrt{2\pi}} \frac{e^{-\frac{1}{2}\left(\gamma + \delta \ln\left(\frac{x-\xi}{\xi+\lambda-x}\right)\right)^2}}{\left(\frac{x-\xi}{\lambda}\right)\left(1 - \frac{x-\xi}{\lambda}\right)} $$

*   **Esperanza (Media):**
    *   No tiene forma cerrada general. Depende del tipo ($S_U$ o $S_B$) y parámetros.

*   **Varianza:**
    *   No tiene forma cerrada general.

*   **Moda:**
    *   Se obtiene resolviendo $f'(x) = 0$ numéricamente.

**Gráfica**

image.png

image.png

**Método seleccionado**

La distribución Johnson se define explícitamente como una transformación de una variable normal estándar $Z∼N(0,1)$, lo que permite generarla de manera exacta y eficiente mediante el método de transformación inversa aplicada a normales.

**Pseudocodigo**

1. Generar $Z ~ N(0,1)$ usando Box-Muller o cualquier otro método exacto.
2. Aplicar la transformación Johnson $ S_U $:
   $$X = ξ + λ * sinh((Z - γ) / δ)$$
3. Retornar X

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def simular_johnson(xi, lamda, gamma, delta, n=1000):
    # Generar Z ~ N(0,1) usando Box-Muller
    U1 = np.random.rand(n)
    U2 = np.random.rand(n)
    Z = np.sqrt(-2 * np.log(U1)) * np.cos(2 * np.pi * U2)

    # Aplicar transformación Johnson SU
    X = xi + lamda * np.sinh((Z - gamma) / delta)
    return X

# Parámetros y simulación
xi = 0.0     # ubicación
lamda = 1.0   # escala
gamma = 0.5   # forma 1
delta = 2.0   # forma 2
muestras = simular_johnson(xi, lamda, gamma, delta)

# Estadísticas muestrales
media_muestral = np.mean(muestras)
varianza_muestral = np.var(muestras)

# Estadísticas teóricas (fórmulas para Johnson SU)
omega = np.exp(1/delta**2)
media_teorica = xi - lamda * np.sqrt(omega) * np.sinh(gamma/delta)
varianza_teorica = (lamda**2)/2 * (omega - 1) * (omega * np.cosh(2*gamma/delta) + 1)

print('La media es:', media_muestral)
print('La varianza es:', varianza_muestral)
print('La media teórica es:', media_teorica)
print('La varianza teórica es:', varianza_teorica)

# Configurar gráficos
plt.figure(figsize=(12, 4))

# Histograma (PDF)
plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='teal', edgecolor='black')
x = np.linspace(np.min(muestras), np.max(muestras), 1000)
# PDF teórica (aproximación)
pdf_teorica = (delta / (lamda * np.sqrt(2*np.pi))) * \
              1/np.sqrt(1 + ((x-xi)/lamda)**2) * \
              np.exp(-0.5 * (gamma + delta * np.arcsinh((x-xi)/lamda))**2)
plt.plot(x, pdf_teorica, 'r-', lw=2)
plt.title(f'PDF Johnson SU(ξ={xi}, λ={lamda}, γ={gamma}, δ={delta})')
plt.xlabel('Valor')
plt.ylabel('Densidad')

# Gráfico CDF empírica vs teórica
plt.subplot(1, 2, 2)
plt.scatter(np.sort(muestras), np.linspace(0, 1, len(muestras)),
            color='green', s=10, label='Empírica')
# CDF teórica (aproximación)
cdf_teorica = stats.norm.cdf(gamma + delta * np.arcsinh((x-xi)/lamda))
plt.plot(x, cdf_teorica, 'r--', label='Teórica')
plt.title('CDF Johnson SU')
plt.xlabel('Valor')
plt.ylabel('Probabilidad Acumulada')
plt.legend()

plt.tight_layout()
plt.show()
```

##Uniforme

**Nomenclatura**

*   **Notación:** $X \sim \text{Uniforme Discreta}(a, b)$
    *   "La variable aleatoria X sigue una distribución uniforme discreta entre $a$ y $b$ (enteros)."

*   **Parámetros:**
    *   $a$: Límite inferior (entero mínimo)
    *   $b$: Límite superior (entero máximo), con $a \leq b$

*   **Soporte:** $\{a, a+1, ..., b-1, b\}$
    *   Todos los enteros entre $a$ y $b$ inclusive

*   **Función de Masa de Probabilidad (PMF):**
    $$ P(X = k) = \frac{1}{b - a + 1} \quad \text{para } k = a, a+1, ..., b $$

*   **Función de Distribución Acumulada (CDF):**
    $$ F(k) = \frac{\lfloor k \rfloor - a + 1}{b - a + 1} \quad \text{para } a \leq k \leq b $$

*   **Esperanza (Media):**
    $$ E(X) = \frac{a + b}{2} $$

*   **Varianza:**
    $$ Var(X) = \frac{(b - a + 1)^2 - 1}{12} $$

*   **Moda:**
    *   Todos los valores son igualmente probables (no única)


**Gráfica**

image.png

image.png

**Método seleccionado**

La uniforme discreta es la distribución más simple para variables enteras. El método de transformada inversa es el más eficiente, ya que solo requiere generar una variable uniforme continua y aplicarle una operación de redondeo/discretización.

**Pseudocodigo**

1. Generar $U \sim $Uniforme$(0, 1)$
2. Calcular $X = a + floor((b - a + 1) * U)$
3. Retornar $X $

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt

def simular_uniforme_discreta(a, b, n=1000):
    U = np.random.rand(n)
    return a + np.floor((b - a + 1) * U)

# Parámetros y simulación
a=1
b=6
muestras = simular_uniforme_discreta(a, b, 10000)

# Cálculos estadísticos
print('La media es:', np.mean(muestras))
print('La varianza es:', np.var(muestras))

print('La media teorica es:', (a + b)/2)
print('La varianza teorica es:', ((b - a + 1)**2 - 1)/12)


# Configurar gráficos
plt.figure(figsize=(12, 5))

# PMF (Función de Masa de Probabilidad)
plt.subplot(1, 2, 1)
valores, counts = np.unique(muestras, return_counts=True)
plt.bar(valores, counts/len(muestras), color='teal',
        edgecolor='black', alpha=0.7, label='Empírica')
plt.axhline(1/(b-a+1), color='red', linestyle='--', label='Teórica')
plt.xticks(np.arange(a, b+1))
plt.title(f'PMF Uniforme Discreta({a}, {b})')
plt.xlabel('Valor')
plt.ylabel('Probabilidad')
plt.legend()

# CDF (Función de Distribución Acumulativa)
plt.subplot(1, 2, 2)
x = np.arange(a-1, b+2)
cdf_empirica = np.array([np.mean(muestras <= v) for v in x])
cdf_teorica = np.array([max(0, min(1, (v-a+1)/(b-a+1))) for v in x])  # Corregido aquí
plt.step(x, cdf_empirica, where='post', color='green', label='Empírica')
plt.step(x, cdf_teorica, where='post', color='red',
         linestyle='--', label='Teórica')
plt.xticks(np.arange(a-1, b+2))
plt.title('CDF Uniforme Discreta')
plt.xlabel('Valor')
plt.ylabel('Probabilidad Acumulada')
plt.legend()

plt.tight_layout()
plt.show()
```

##Bernoulli

**Nomenclatura**

*   **Notación:** $X \sim \text{Bernoulli}(p)$
    *   "La variable aleatoria X sigue una distribución Bernoulli con parámetro $p$."

*   **Parámetro:**
    *   $p$: Probabilidad de éxito ($0 \leq p \leq 1$)

*   **Soporte:** $\{0, 1\}$
    *   0: "Fracaso"
    *   1: "Éxito"

*   **Función de Masa de Probabilidad (PMF):**
    $$ P(X = k) = \begin{cases}
    p & \text{si } k = 1 \\
    1-p & \text{si } k = 0
    \end{cases} $$
    *   Forma compacta: $p^k(1-p)^{1-k}$ para $k \in \{0,1\}$

*   **Función de Distribución Acumulada (CDF):**
    $$ F(k) = \begin{cases}
    0 & \text{para } k < 0 \\
    1-p & \text{para } 0 \leq k < 1 \\
    1 & \text{para } k \geq 1
    \end{cases} $$

*   **Esperanza (Media):**
    $$ E(X) = p $$

*   **Varianza:**
    $$ Var(X) = p(1-p) $$

*   **Moda:**
    $$ \text{Moda} = \begin{cases}
    0 & \text{si } p < 0.5 \\
    1 & \text{si } p > 0.5 \\
    \text{ambos} & \text{si } p = 0.5
    \end{cases} $$

La distribución Bernoulli es la más simple para variables binarias (0 o 1). El método de **transformada inversa** es el más eficiente, ya que solo requiere generar una variable uniforme continua y compararla con el parámetro $p$



**Pseudocodigo**

3. Retornar X

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt

def simular_bernoulli(p, n=1000):
    U = np.random.rand(n)
    return (U <= p)  # Devuelve booleanos (True=1, False=0 en cálculos numéricos)

# Parámetros y simulación
p = 0.3
muestras = simular_bernoulli(p, 10000)

#Calculos
print('La media es:', np.mean(muestras))
print('La varianza es:', np.var(muestras))

print('La media teorica es:', p)
print('La varianza teorica es:', p * (1 - p))

# Gráficos (igual que antes)
plt.figure(figsize=(12, 5))

# PMF
plt.subplot(1, 2, 1)
valores, counts = np.unique(muestras, return_counts=True)
plt.bar(valores, counts / len(muestras), color='teal', edgecolor='black', alpha=0.7, label='Empírica')
plt.scatter([0, 1], [1 - p, p], color='red', zorder=5, label='Teórica')
plt.xticks([0, 1])
plt.title(f'PMF Bernoulli(p={p})')
plt.xlabel('Valor')
plt.ylabel('Probabilidad')
plt.legend()

# CDF
plt.subplot(1, 2, 2)
x = np.linspace(-1, 2, 1000)
cdf_empirica = np.array([np.mean(muestras <= v) for v in x])
cdf_teorica = np.piecewise(x, [x < 0, (0 <= x) & (x < 1), x >= 1], [0, lambda x: 1 - p, 1])
plt.step(x, cdf_empirica, color='teal', label='Empírica',linewidth=2.5)
plt.step(x, cdf_teorica, color='red', linestyle='--', label='Teórica',linewidth=2.5)
plt.xticks([-1, 0, 1, 2])
plt.title('CDF Bernoulli')
plt.xlabel('Valor')
plt.ylabel('Probabilidad Acumulada')
plt.legend()

plt.tight_layout()
plt.show()
```

##Geométrica

**Nomenclatura**

*   **Notación:** $X \sim \text{Geom}(p)$
    *   "La variable aleatoria X sigue una distribución geométrica con parámetro $p$."

*   **Parámetro:**
    *   $p$: Probabilidad de éxito en cada ensayo ($0 < p \leq 1$)

*   **Soporte:** $\{1, 2, 3, \dots\}$ (versión para número de ensayos hasta el primer éxito)
    *   Alternativamente: $\{0, 1, 2, \dots\}$ (para número de fracasos antes del primer éxito)

*   **Función de Masa de Probabilidad (PMF):**
    $$ P(X = k) = (1-p)^{k-1}p \quad \text{para } k = 1, 2, 3, \dots $$

*   **Función de Distribución Acumulada (CDF):**
    $$ F(k) = 1 - (1-p)^k \quad \text{para } k \geq 0 $$

*   **Esperanza (Media):**
    $$ E(X) = \frac{1}{p} $$

*   **Varianza:**
    $$ Var(X) = \frac{1-p}{p^2} $$

*   **Moda:**
    $$ \text{Moda} = 1 $$

**Gráfica**

**Método seleccionado**

Transformada Inversa usando Exponenciales
La distribución geométrica modela el número de intentos hasta el primer éxito en ensayos Bernoulli. El método más eficiente y exacto aprovecha la relación entre la geométrica y la exponencial mediante:
$$X=[ \frac{In(1-U)}{In(1-p)}]$$

donde $U\sim$ Uniforme $(0,1)$. Este método es exacto y solo requiere generar una uniforme.

**Pseudocodigo**

1. Generar $U \sim$Uniforme$(0, 1)$
2. Calcular $X =$ techo$( ln(1 - U) / ln(1 - p) )$
3. Retornar $X$

*La función techo es para el primer entero mayor o igual*

**Código**

```python
import numpy as np
import math

def simular_geometrica(p, size=1):
    """
    Simula variables Geométrica(p).
    - p: probabilidad de éxito (0 < p < 1).
    - size: número de muestras.
    """
    U = np.random.uniform(size=size)
    X = np.ceil(np.log(1 - U) / np.log(1 - p)).astype(int)
    return X


# Parámetro: p = 0.2 (20% de probabilidad de éxito por intento)
p = 0.2

# Generar 10,000 muestras
muestras = simular_geometrica(p, 10_000)

# Estadísticas descriptivas
media_muestral = np.mean(muestras)
print(f"Media teórica (1/p): {1/p:.1f}")
print(f"Media muestral: {media_muestral:.1f}")

# Visualización
import matplotlib.pyplot as plt
plt.hist(muestras, bins=range(1, 21), density=True, alpha=0.7, align='left')
plt.title(f'Distribución Geométrica(p={p})')
plt.xlabel('Número de intentos hasta el primer éxito')
plt.ylabel('Probabilidad')
plt.show()
```

##Binomial negativa

**Nomenclatura**

*   **Notación:**$ X \sim \text{NB}(r, p) $
    *   "La variable aleatoria $\( X \)$ sigue una distribución binomial negativa con parámetros $\( r \)$ y $\( p \)$."

*   **Parámetros:**
    *   $ r$: Número de éxitos requeridos (entero positivo, $ r > 0 $)
    *   $ p $: Probabilidad de éxito en cada ensayo ($ 0 < p \leq 1 $)

*   **Soporte:** $ \{0, 1, 2, \dots\}$
    *   Número de fracasos antes del \( r \)-ésimo éxito

*   **Función de Masa de Probabilidad (PMF):**
    $$ P(X = k) = \binom{k + r - 1}{r - 1} p^r (1 - p)^k \quad \text{para } k = 0, 1, 2, \dots $$

*   **Función de Distribución Acumulada (CDF):**
    $$ F(k) = I_p(r, k + 1) $$
    *   Donde $ I_p $ es la función beta incompleta regularizada

*   **Esperanza (Media):**
    $$ E(X) = \frac{r(1 - p)}{p} $$

*   **Varianza:**
    $$ \text{Var}(X) = \frac{r(1 - p)}{p^2} $$

*   **Moda:**
    $$ \text{Moda} = \left\lfloor \frac{(r - 1)(1 - p)}{p} \right\rfloor \quad \text{para } r > 1 $$

**Gráfica**

image.png

image.png

**Método seleccionado**

Para simular una Binomial Negativa (discreta), el mejor método entre tus opciones es composición, ya que podemos aprovechar su relación con la distribución Gamma y Poisson.

**Pseudocodigo**

1. Generar $Y \sim \text{Gamma}\left(r, \frac{1-p}{p}\right)$:

   1.1. Usar $r$ como parámetro de forma

   1.2. Usar $\frac{1-p}{p}$ como parámetro de escala

2. Generar $X \sim \text{Poisson}(Y)$:

   2.1. Usar $Y$ como parámetro de tasa

3. Retornar $X \sim \text{BinNeg}(r, p)$

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom

def simular_binneg(r, p, n=1000):
    # Generar Y ~ Gamma(r, (1-p)/p)
    Y = np.random.gamma(shape=r, scale=(1-p)/p, size=n)

    # Generar X ~ Poisson(Y)
    X = np.random.poisson(lam=Y)

    return X

# Parámetros y simulación
r, p = 5, 0.3  # 5 éxitos, probabilidad 0.3
muestras = simular_binneg(r, p, 10000)  # 10,000 muestras


# Cálculos estadísticos

teorica_media = r * (1-p) / p
teorica_varianza = r * (1-p) / p**2

print('Media muestral:', np.mean(muestras))
print('Varianza muestral:', np.var(muestras))
print('\nMedia teórica:', teorica_media)
print('Varianza teórica:', teorica_varianza)


# Gráficos

plt.figure(figsize=(12, 5))

# Histograma (PMF empírica)
plt.subplot(1, 2, 1)
counts, bins, _ = plt.hist(muestras, bins=range(0, 40), density=True,
                           color='skyblue', edgecolor='black', alpha=0.7,
                           label='Empírica')

# PMF teórica
x = np.arange(0, 40)
pmf_teorica = nbinom.pmf(x, r, p)
plt.stem(x, pmf_teorica, linefmt='r-', markerfmt='ro',
         basefmt=' ', label='Teórica')
plt.title(f'PMF Binomial Negativa(r={r}, p={p})')
plt.xlabel('Valor')
plt.ylabel('Probabilidad')
plt.legend()

# CDF empírica vs teórica
plt.subplot(1, 2, 2)
cdf_empirica = np.cumsum(np.bincount(muestras.astype(int)) / len(muestras))
cdf_teorica = nbinom.cdf(x, r, p)
plt.step(x, cdf_empirica[:40], 'b-', where='post', label='Empírica')
plt.step(x, cdf_teorica, 'r--', where='post', label='Teórica')
plt.title(f'CDF Binomial Negativa(r={r}, p={p})')
plt.xlabel('Valor')
plt.ylabel('Probabilidad Acumulada')
plt.legend()

plt.tight_layout()
plt.show()
```

##Poisson

**Nomenclatura**

*   **Notación:** $ X \sim \text{Poisson}(\lambda) $
    *   "La variable aleatoria \( X \) sigue una distribución de Poisson con parámetro $ \lambda $."

*   **Parámetro:**
    *   $ \lambda $: Tasa de ocurrencia (valor esperado de eventos), $\( \lambda > 0 \)$

*   **Soporte:** $\{0, 1, 2, \dots\} $
    *   Todos los enteros no negativos

*   **Función de Masa de Probabilidad (PMF):**
    $$ P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!} \quad \text{para } k = 0, 1, 2, \dots $$

*   **Función de Distribución Acumulada (CDF):**
    $$ F(k) = e^{-\lambda} \sum_{i=0}^{\lfloor k \rfloor} \frac{\lambda^i}{i!} $$

*   **Esperanza (Media):**
    $$ E(X) = \lambda $$

*   **Varianza:**
    $$ \text{Var}(X) = \lambda $$
    *   Propiedad única: media = varianza

*   **Moda:**
    $$ \text{Moda} = \lfloor \lambda \rfloor \quad \text{y} \quad \lceil \lambda \rceil - 1 \quad \text{si } \lambda \text{ es entero} $$


**Gráfica**

image.png

image.png

**Método seleccionado**

El método de **aceptación-rechazo** es ideal para Poisson cuando se necesita precisión para cualquier valor de $\lambda$ (pequeño o grande), sin depender de aproximaciones. Usaremos una distribución envolvente basada en exponenciales para mayor eficiencia.

**Pseudocodigo**

1. Calcular c = sqrt(2πλ) * e^λ / λ^λ
2. Calcular p = λ/(λ + 1)
3. Generar Y ~ Geométrica(p) (devuelve Y ∈ {1,2,...})
4. Restar 1: Y = Y - 1 (ahora Y ∈ {0,1,2,...})
5. Generar U ~ Uniforme(0,1)
6. Calcular probabilidad_aceptar = (e^λ * λ^Y) / (c * Y!)
7. Si U ≤ probabilidad_aceptar:
     Retornar X = Y  (Aceptado)
   Si no:
     Volver al paso 3 (Rechazado)

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
from math import exp, factorial

def simular_poisson(lambd, n=1000):

    muestra = np.zeros(n, dtype=int)
    L = exp(-lambd)

    for i in range(n):
        k = 0
        p = 1.0
        while True:
            k += 1
            u = np.random.rand()
            p *= u
            if p <= L:
                break
        muestra[i] = k - 1

    return muestra

# Parámetros y simulación
lambd = 4.5  # Parámetro de la Poisson
muestras = simular_poisson(lambd, 10000)


# Cálculos estadísticos

print('La media muestral es:', np.mean(muestras))
print('La varianza muestral es:', np.var(muestras))

print('\nLa media teórica es:', lambd)
print('La varianza teórica es:', lambd)

# Gráficos

plt.figure(figsize=(12, 5))

# PMF (Función de Masa de Probabilidad)
plt.subplot(1, 2, 1)
valores, counts = np.unique(muestras, return_counts=True)
plt.bar(valores, counts/len(muestras), color='skyblue',
        edgecolor='black', alpha=0.7, label='Empírica')

# PMF teórica
k = np.arange(0, max(valores)+1)
pmf_teorica = np.exp(-lambd) * (lambd**k) / np.vectorize(factorial)(k)
plt.scatter(k, pmf_teorica, color='red', zorder=5, label='Teórica')
plt.title(f'PMF Poisson(λ={lambd})')
plt.xlabel('Valor')
plt.ylabel('Probabilidad')
plt.legend()

# CDF (Función de Distribución Acumulativa)
plt.subplot(1, 2, 2)
x = np.arange(-1, max(valores)+2)
cdf_empirica = np.array([np.mean(muestras <= v) for v in x])
cdf_teorica = np.cumsum(pmf_teorica)
cdf_teorica_extendida = np.concatenate(([0], cdf_teorica))
x_teorico = np.arange(-1, len(pmf_teorica))
plt.step(x, cdf_empirica, where='post', color='blue', label='Empírica')
plt.step(x_teorico, cdf_teorica_extendida,
         where='post', color='red', linestyle='--', label='Teórica')
plt.title('CDF Poisson')
plt.xlabel('Valor')
plt.ylabel('Probabilidad Acumulada')
plt.legend()

plt.tight_layout()
plt.show()
```
