**Problema**

Dos personas han quedado citas en un determinado lugar. Ambas tienen la misma
probabilidad de llegar al lugar de la cita en el intervalo de tiempo [0, 30] (minutos).
Determínese la probabilidad de que el tiempo que una persona espere a la otra sea
como máximo 10 minutos.

```python
import matplotlib.pyplot as plt
import random

def simulacion(n=10000):
    exitos = 0
    valido = []
    invalido = []
    prob_historico = []

    for i in range(1, n+1):
        x, y = random.uniform(0, 30), random.uniform(0, 30)
        if abs(x - y) <= 10:
            exitos += 1
            valido.append((x, y))
        else:
            invalido.append((x, y))

        # Guardar probabilidad cada 100 iteraciones
        if i % 100 == 0:
            prob_historico.append(exitos / i)

    prob = exitos / n
    print(f"Probabilidad estimada: {prob:.4f} ")

    # Gráfico de dispersión
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.scatter(*zip(*valido), color='teal', alpha=0.5, label='≤ 10 min', s=10)
    plt.scatter(*zip(*invalido), color='black', alpha=0.5, label='> 10 min', s=10)
    plt.plot([0, 30], [10, 40], 'k--', [0, 30], [-10, 20], 'k--', linewidth=1)
    plt.title('Tiempos de llegada')
    plt.xlabel('Persona 1 (min)')
    plt.ylabel('Persona 2 (min)')
    plt.xlim(0, 30)
    plt.ylim(0, 30)
    plt.legend()
    plt.grid(True)

    # Gráfico de evolución
    plt.subplot(1, 2, 2)
    plt.plot(range(100, n+1, 100), prob_historico, color='blue', alpha=0.7)
    plt.axhline(y=5/9, color='red', linestyle='--', label='Valor teórico')
    plt.title('Probabilidad')
    plt.xlabel('Número de simulaciones')
    plt.ylabel('Probabilidad')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

simulacion()
```
