**Método Box-Müller**



El método Box-Müller es una técnica para generar pares de números aleatorios independientes que siguen una distribución normal estándar (media 0, varianza 1), a partir de números aleatorios con distribución uniforme en el intervalo (0, 1).



**Idea central**

La idea central del método se basa en una transformación de variables aleatorias. Se parte de dos variables aleatorias independientes, U1 y U2, que siguen una distribución uniforme en (0, 1). Luego, mediante una transformación, se obtienen dos nuevas variables, X e Y, que siguen una distribución normal estándar.



**La transformación**

La transformación se realiza en dos pasos:

a) Coordenadas Polares: Se convierten U1 y U2 a coordenadas polares R y

Theta usando las siguientes fórmulas:

1. R = sqrt(-2 * ln(U1))

2. Theta = 2 * pi * U2

R representa la distancia al origen en el plano cartesiano.

Theta representa el ángulo con respecto al eje horizontal.

b) Coordenadas Cartesianas: Se convierten las coordenadas polares R y Theta a coordenadas cartesianas X e Y usando las siguientes fórmulas:



1.   X = R * cos(Theta)
2.   Y = R * sin(Theta)

X e Y son las variables que seguirán la distribución normal estándar.



**En términos simples:**

Imagina que lanzas dardos aleatoriamente a un tablero circular. La posición de cada dardo se puede describir con coordenadas polares (distancia al centro, ángulo). El método Box-Müller es como si convirtiéramos la posición de los dardos en coordenadas cartesianas (x, y). Sorprendentemente, la distribución de los dardos en el plano cartesiano (x, y) se asemeja a una distribución normal, incluso si los dardos fueron lanzados de forma uniforme al tablero circular.

**Argumentación del código**

1.   Importar librerías: Se importan las librerías numpy, matplotlib.pyplot y random para realizar cálculos numéricos, generar gráficos y obtener números aleatorios, respectivamente.
2.  Función box: Esta función implementa el método Box-Müller. Recibe como entrada el número de muestras a generar (n). Dentro de un bucle, se generan dos números aleatorios uniformes U1 y U2. Luego, se calculan R, Theta, X e Y utilizando las fórmulas del teorema de Box-Müller. Los valores de X e Y se almacenan en las listas listax y listay, respectivamente. Finalmente, la función devuelve estas dos listas.
3. Generar muestras: Se llama a la función box con un argumento de 100000 para generar 100000 muestras. Los resultados se almacenan en la variable experimentos.
4. Calcular media y varianza: Se calcula la media y la varianza de las muestras generadas utilizando las funciones np.mean y np.var, respectivamente. Estos valores se imprimen en la consola.
5. Generar histograma: Se genera un histograma de las muestras utilizando la función plt.hist. El histograma se ajusta a una curva de densidad para visualizar la distribución de las muestras. Se agrega una cuadrícula, un título y se muestra el gráfico.








```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd

#Generamos nuestra funcion

listax=[]
listay=[]
def box(n):
  for _ in range(n):
    U1=rd.uniform(0,1)
    U2=rd.uniform(0,1)
    R=np.sqrt(-2*np.log(U1))
    Theta=2*np.pi*U2
    X=R*np.cos(Theta)
    Y=R*np.sin(Theta)
    listax.append(X)
    listay.append(Y)
  return listax,listay

experimentos=box(100000)

print('la media es', np.mean(experimentos))
print('la varianza es', np.var(experimentos))


plt.hist(experimentos, bins=100, density=True)
plt.grid(True, linestyle='--', alpha=0.7)
plt.title('Histograma de la distribución')
plt.show()
```
