**Numeros aleatorios**

Generador Congruencial \
*$x_{n+1} =  (a*x_n + c) \mod m $ * \\
${x_0}, a, c, m$ son constantes enteras.

```python
import tabulate as tb

# Parámetros para el generador lineal congruente
a = 101 #Constante multiplicativo
c = 1021 #Constante de incremento
m = 100 #Modulo
x_0 = 3 #Valor inicial
n = 110 #Número de iteraciones
muestra = []

# Función generadora
def generador(a, c, m, x_0):
    return (a * x_0 + c) % m


# Imprime las n iteraciones de la función
#for _ in range(n):
    #x_0 = generador(a, c, m, x_0)
    #print(x_0)

# Generación de tabla con Tabulate
x_0 = 3  # Reiniciamos el valor inicial
muestra_tab = []
valores_anterior = {}  # Usamos un diccionario para rastrear la posición de los valores generados

for i in range(n):
    nuevo = generador(a, c, m, x_0)
    # Si el valor ya ha sido generado, calculamos el ciclo de vida
    if nuevo in valores_anterior:
        ciclo_vida = i - valores_anterior[nuevo]  # El ciclo de vida es la diferencia entre las posiciones
    else:
        ciclo_vida = "Generación"

    valores_anterior[nuevo] = i  # Guardamos la posición del valor generado

    muestra_tab.append([i, x_0, nuevo, nuevo/m])

    x_0 = nuevo

encabezados = ["i", "X_i", "X_i+1","X_i+1 (normalizado)"]
print(tb.tabulate(muestra_tab, headers=encabezados, tablefmt="fancy_grid"))

# Imprimir mensaje final indicando el ciclo de vida
print("\nCiclo de vida:", ciclo_vida)

```
