**Numeros aleatorios**

Generador Congruencial \
*$x_{n+1} =  (a*x_n + c) \mod m $ * \\
${x_0}, a, c, m$ son constantes enteras.

```python
import pandas as pd
import tabulate as tb

# Parámetros
a = 16
c = 203
m = 22
x_0 = 3
n = 30
muestra = []

# Función generadora
def generador(a, c, m, x_0):
    return (a * x_0 + c) % m

# Imprime una iteración
print(generador(a, c, m, x_0))

# Imprime las n iteraciones de la función
for _ in range(n):
    x_0 = generador(a, c, m, x_0)
    print(x_0)

# Generación y almacenamiento de valores en un DataFrame
x_0 = 3  # Reiniciamos el valor inicial
for i in range(n):
    nuevo = generador(a, c, m, x_0)
    muestra.append([i, x_0, nuevo, nuevo / m])
    x_0 = nuevo

df = pd.DataFrame(muestra, columns=["Iteración", "Xn", "X_n+1", "Xn/m"])
print(df)

# Generación de tabla con Tabulate
x_0 = 3  # Reiniciamos el valor inicial
muestra_tab = []
for i in range(n):
    nuevo = generador(a, c, m, x_0)
    muestra_tab.append([i, x_0, nuevo])
    x_0 = nuevo

encabezados = ["i", "X_i", "X_i+1"]
print(tb.tabulate(muestra_tab, headers=encabezados, tablefmt="fancy_grid"))
```
