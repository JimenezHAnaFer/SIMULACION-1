**Numeros aleatorios**

Generador Congruencial \
*$x_{n+1} =  (a*x_n + c) \mod m $ * \\
${x_0}, a, c, m$ son constantes enteras.

```python
import pandas as pd
import tabulate as tb

# Parámetros
a = 16
c = 203
m = 22
x_0 = 3
n = 30
muestra = []

# Función generadora
def generador(a, c, m, x_0):
    return (a * x_0 + c) % m

# Imprime una iteración
print(generador(a, c, m, x_0))

# Imprime las n iteraciones de la función
for _ in range(n):
    x_0 = generador(a, c, m, x_0)
    print(x_0)

# Generación y almacenamiento de valores en un DataFrame
x_0 = 3  # Reiniciamos el valor inicial
for i in range(n):
    nuevo = generador(a, c, m, x_0)
    muestra.append([i, x_0, nuevo, nuevo / m])
    x_0 = nuevo

df = pd.DataFrame(muestra, columns=["Iteración", "Xn", "X_n+1", "Xn/m"])
print(df)

# Generación de tabla con Tabulate
x_0 = 3  # Reiniciamos el valor inicial
muestra_tab = []
for i in range(n):
    nuevo = generador(a, c, m, x_0)
    muestra_tab.append([i, x_0, nuevo])
    x_0 = nuevo

encabezados = ["i", "X_i", "X_i+1"]
print(tb.tabulate(muestra_tab, headers=encabezados, tablefmt="fancy_grid"))
```

```python
import tabulate
import matplotlib.pyplot as plt
from numpy import var
import numpy as np

a = 1001; # cambiar valores a y c
c = 1021;
m = 100**2+1;
x0 = 5;
n = m;

#Función generadora

def generador(a, c, m, x0):
  x = (a*x0 + c) % m
  return x

muestra =[]
lista=[] #guardamos los datos normalizados

for i in range(n):
  nuevo = generador(a,c,m,x0)
  muestra.append([i,x0,nuevo/m])
  lista.append(nuevo/m)
  x0=nuevo
encabezados=["i","x_1","x_{i+1}"]
print(tabulate.tabulate(muestra,headers = encabezados))

#Varianza y media
vari= var(lista)
print("Varianza", vari) #Imprimir el valor de la varianza
mean = np.mean(lista)
print("Media", mean) #Imprimir el valor de la media





#Gráfica
plt.plot(lista);
plt.ylim(0,1.5);
plt.show();
#histograma
plt.hist(lista);
plt.show();

#probabilidad acumulada
lista_sort=np.sort(lista); #acomodar valores de la lista
print(lista_sort);
y=np.arange(1, m+1)/m;
plt.plot(lista_sort,y);
plt.show();
```
