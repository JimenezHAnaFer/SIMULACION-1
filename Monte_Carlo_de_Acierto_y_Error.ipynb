# Monte Carlo de Acierto y Error

El método Monte Carlo es una técnica computacional que utiliza muestreo aleatorio para obtener resultados numéricos. En su variante de "acierto y error" (o "aceptación-rechazo"),
La simulación de Montecarlo es una de las herramientas más potentes en el análisis de incertidumbre y toma de decisiones estratégicas. Este método se usa ampliamente en diversas industrias para modelar escenarios futuros y evaluar riesgos en función de datos probabilísticos.







**¿Cómo funciona la simulación de Monte Carlo?**

El proceso de la simulación de Montecarlo sigue una serie de pasos clave:

1. Definición del problema: Se identifican las variables clave que afectan el sistema a analizar.
2. Asignación de distribuciones de probabilidad: Cada variable se modela con una distribución como la normal, uniforme o triangular.
3. Generación de números aleatorios: Se crean valores simulados para las variables utilizando modelos probabilísticos.
4. Ejecutar múltiples simulaciones: Se realizan multiples iteraciones para calcular los posibles resultados.
5. Análisis de resultados: Se obtiene una distribución final con la probabilidad de cada posible desenlace.


# **Código**

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

# Definimos nuestra función
def f(x):
    return np.sqrt(np.arctan(x))

x = np.linspace(0, 1, 10000)
y = f(x)

#print(x)
#print(y)

#Definimos la función de Monte Carlo
def dardos(n):
    exitos = 0  # Contador
    for _ in range(n):
        u1 = rd.random()
        u2 = rd.random()
        if u2 <= f(u1):  # Si está debajo de la curva, es un éxito
            exitos += 1
    return exitos / n

dardos(1000)
# Calcular probabilidad
probabilidad = dardos(1000000)
print('La probabilidad es:',probabilidad)

# Graficar la función
plt.plot(x, y)
plt.axhline(f(1), color="red")
plt.axvline(1, color="red")

# Graficar puntos aleatorios de prueba
u1 = rd.random()
u2 = rd.random()
plt.scatter(u1, u2, color="green", label="Ejemplo de punto aleatorio")



plt.scatter(u1, u2, color = "green")
plt.show()


# Función para realizar múltiples experimentos
def experimento(m, n=10000):
    lista = [dardos(n) for _ in range(m)]
    return lista

# Ejecutamos el experimento 300 veces
l = experimento(300)
print("Última estimación:", l[-1])


# Graficamos las estimaciones
plt.plot(l)
plt.axhline(np.mean(l), color= "red" , linestyle= "--" )
plt.axhline(np.mean(l) + np.std(l), color= "green" , linestyle= "--" )
plt.axhline(np.mean(l) - np.std(l), color= "green" , linestyle= "--" )
plt.legend()
plt.show()

# Resultados
print ( "Media de las estimaciones:" , np.mean(l))
print ( "Varianza de las estimaciones:" , np.var(l))
```
